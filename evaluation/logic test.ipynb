```python
def openai_get3_call(context, query):

  prompt = '''Find out what action enables the given action. Answer none if there is no connection.

Plot: John walked to get to store.<\s> John buys a uniform from a shop.<\s> John took a bus to get to school.<\s> John did karate in high school at school<\s> John went to get to store.<\s> John buys a wrestling uniform at a store.<\s> John took a bus to get to laundry.<\s> John washed the wrestling uniform at laundry<\s>John wrestled to get to wrestling.<\s> John became a pro wrestler at wrestling.
Question: What's the action, if any, enabled John buys a uniform from a shop?
Answer: John walked to get to store.
Question: What's the action, if any, enabled John took a bus to get to school?
Answer: none.
Question: What's the action, if any, enabled John did karate in high school at school?
Answer: John took a bus to get to school.
Question: What's the action, if any, enabled John went to get to store?
Answer: none.
Question: What's the action, if any, enabled John buys a wrestling uniform at a store?
Answer: John went to get to store.
Question: What's the action, if any, enabled John took a bus to get to laundry?
Answer: John buys a wrestling uniform at a store.
Question: What's the action, if any, enabled John washed the wrestling uniform at laundry?
Answer: John buys a wrestling uniform at a store.

Plot: Tom was having chest pains.<\s> He didn't know what was causing them.<\s> He went to the doctor.<\s> It turned out to be indigestion.<\s> Tom was relieved it wasn't something worse.
Question: What's the action, if any, enabled He didn't know what was causing them?
Answer: none.
Question: What's the action, if any, enabled He went to the doctor?
Answer: Tom was having chest pains.
Question: What's the action, if any, enabled It turned out to be indigestion?
Answer: none.
Question: What's the action, if any, enabled Tom was relieved it wasn't something worse?
Answer: Tom was having chest pains.

Plot: John took a bus to get to restaurant.<\s>John walked to get to hospital.<\s> John used his money to buy stuff at shop.<\s> John bought bias from the garden.<\s> John wrote a school paper at school.
Question: What's the action, if any, enabled John walked to get to hospital?
Answer: none.
Question: What's the action, if any, enabled John used his money to buy stuff at shop?
Answer: none.
Question: What's the action, if any, enabled John bought bias from the garden?
Answer: none.
Question: What's the action, if any, enabled John wrote a school paper at school?
Answer: none.

Plot: mike had a dog. he felt bad for it. he decided to buy a new dog. he went to the pet store. he bought a new dog. 
Question: What's the action, if any, enabled he felt bad for it?
Answer: none.
Question: What's the action, if any, enabled he decided to buy a new dog?
Answer: none.
Question: What's the action, if any, enabled he went to the pet store?
Answer: he decided to buy a new dog.
Question: What's the action, if any, enabled he bought a new dog?
Answer: he went to the pet store.

Plot: {$text1$}
Question: What's the action, if any, enabled {$text2$}?
Answer:'''
  prompt = prompt.replace("{$text1$}",context.strip()).replace("{$text2$}",query.strip())
  #print(prompt)
  response = openai.Completion.create(
    model="text-davinci-002",
    prompt=prompt,
    temperature=0.7,
    max_tokens=20,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0
  )
  return response.choices[0].text
```

# enablement %

```python
def ask_questions(s):
    sentences = s.split('<\\s>')
    ret = {}
    unanswerable = 0
    total = 0
    repitition = 0
    lookup = [input_string.split() for input_string in sentences]
    #print(lookup)
    for s_ in sentences[1:]:
        total+=1
        res = openai_get3_call('<\\s>'.join(sentences[:sentences.index(s_)+1]),s_).split()
        #print(res)
        if len(set(res).intersection(set(lookup[sentences.index(s_)])))/len(set(lookup[sentences.index(s_)]))>0.7 or all([len(set(res).intersection(set(t)))/len(set(t))<0.2 for t in lookup[:sentences.index(s_)]]):
            ret[s_]=res,2
        else:
            ret[s_]=res,0 if ('none' in res or 'none.' in res) else 1
        if ret[s_][1]==0 or ret[s_][1]==2:
            unanswerable+=1
            if ret[s_][1]==2:
                repitition+=1
        
    return ret,unanswerable,total,repitition
```

```python
import json
results = {}
cum_unanswerable = 0
cum_total = 0
cum_repetiion = 0
file_name = './summary_expanded_with_why_fixed_location.json'
with open(file_name) as json_file:
    plans = json.load(json_file)
    n = 0
    for k in list(plans.keys())[:50]:
        results[k] = {}
        results[k]['unanswerable'] = 0
        results[k]['total'] = 0
        results[k]['repetition'] = 0
        results[k]['detail'] = {}
        res, results[k]['unanswerable'], results[k]['total'], results[k]['repetition'] = ask_questions(plans[k])
        cum_unanswerable+=results[k]['unanswerable']
        cum_total += results[k]['total']
        cum_repetiion += results[k]['repetition'] 
        results[k]['detail'] = res
        print(k,cum_unanswerable/cum_total, cum_repetiion/cum_total)
        
        
        
print(cum_unanswerable/cum_total)
 
```
