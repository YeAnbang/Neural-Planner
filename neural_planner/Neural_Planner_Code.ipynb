```python
from nltk import probability
def infer(context, top_p=None, top_k=None, temp=1.0, gen_len=12, return_probability=False):
    '''
    GPT-J interface:
        returning a generated sentence if return_probability is False
        returning samples, probability, tokens, logits if return_probability is True. 
            samples is the returned sentence in text by default sampling method. (in this colab demo, batch size is 1, so only one sentence is returned)
            probability is a list of probability for each word in samples
            tokens is a list of tokens for each word in samples
            logits: list of numpy vector. posterior probability at each step given the previous tokens. It is probability not logits!!!!
    param context: prompt
    param top_k (int, optional, defaults to model.config.top_k or 50 if the config does not set any value) — The number of highest probability vocabulary tokens to keep for top-k-filtering.
    param top_p (float, optional, defaults to model.config.top_p or 1.0 if the config does not set any value) — If set to float < 1, only the most probable tokens with probabilities that add up to top_p or higher are kept for generation.
    param temp:  (float, optional, defaults to model.config.temperature or 1.0 if the config does not set any value) — The value used to module the next token probabilities.
    param gen_len: num of tokens to generate
    param return_probability: if set to True, return (samples, probability, tokens, logits)
    '''
    tokens = tokenizer.encode(context)

    provided_ctx = len(tokens)
    pad_amount = seq - provided_ctx

    padded_tokens = np.pad(tokens, ((pad_amount, 0),)).astype(np.uint32)
    batched_tokens = np.array([padded_tokens] * total_batch)
    length = np.ones(total_batch, dtype=np.uint32) * len(tokens)

    if top_p and top_k:
        output = network.generate(batched_tokens, length, gen_len,{"top_p": np.ones(total_batch) * top_p, "top_k":np.ones(total_batch) * top_k, "temp": np.ones(total_batch) * temp}, return_logits=return_probability)
    else:
        output = network.generate(batched_tokens, length, gen_len,{"temp": np.ones(total_batch) * temp}, return_logits=return_probability)
    # output = model.generate(batched_tokens, length, gen_len, {"top_p": np.ones(total_batch) * top_p, "temp": np.ones(total_batch) * temp})

    samples = []
    decoded_tokens = output[1][0]

    for o in decoded_tokens[:, :, 0]:
      #samples.append(f"\033[1m{context}\033[0m{tokenizer.decode(o)}")
      samples.append(tokenizer.decode(o))

    #print(f"completion done in {time.time() - start:06}s")
    if return_probability:
        probability = []
        tokens = []
        logits = []
        for i in range(output[1][0].shape[1]):
            p = jax.nn.softmax(output[1][2], axis=- 1)[0][i][0]
            probability.append(float(np.array(p[output[1][0][0][i][0]])))
            tokens.append([int(np.array(output[1][0][0][i][0])),tokenizer.decode(output[1][0][0][i])])
            logits.append(np.array(p))
        return samples, probability, tokens, logits
    return samples

samples,probability, token, logits = infer("John need to give Sally a book, John needs",top_p=0.5,top_k=3,temp=1.0,gen_len=3, return_probability=True)
print(samples)
print(probability)

samples,probability, token, logits = infer("John need to give Sally a book, John needs",top_p=0.5,top_k=3,temp=1.0,gen_len=3, return_probability=True)
print(samples)
print(probability)

samples,probability, token, logits = infer("John need to give Sally a book, John needs",top_p=0.5,top_k=3,temp=1.0,gen_len=3, return_probability=True)
print(samples)
print(probability)

samples,probability, token, logits = infer("John need to give Sally a book, John needs",top_p=0.5,top_k=3,temp=1.0,gen_len=3, return_probability=True)
print(samples)
print(probability)

samples,probability, token, logits = infer("John need to give Sally a book, John needs",top_p=0.5,top_k=3,temp=1.0,gen_len=3, return_probability=True)
print(samples)
print(probability)

samples,probability, token, logits = infer("John need to give Sally a book, John needs",top_p=0.5,top_k=3,temp=1.0,gen_len=3, return_probability=True)
print(samples)
print(probability)
```

```python
def gpt_call(prompt,max_token,num_of_sentences,temperature=0.9, top_p=None, top_k=None, return_probability=False):
    '''
    GPT-J API:
        returning multiple generated sentence if return_probability is False
        returning list of (samples, probability, tokens, logits) if return_probability is True. 
            samples is the generated sentence in text by default sampling method. (in this colab demo, batch size is 1, so only one sentence is returned)
            probability is a list of probability for each word in samples
            tokens is a list of tokens for each word in samples
            logits: list of numpy vector. posterior probability at each step given the previous tokens. It is probability not logits!!!!
    param context: prompt
    param num_of_sentences: number of sentence to generate
    param top_k (int, optional, defaults to model.config.top_k or 50 if the config does not set any value) — The number of highest probability vocabulary tokens to keep for top-k-filtering.
    param top_p (float, optional, defaults to model.config.top_p or 1.0 if the config does not set any value) — If set to float < 1, only the most probable tokens with probabilities that add up to top_p or higher are kept for generation.
    param temperature:  (float, optional, defaults to model.config.temperature or 1.0 if the config does not set any value) — The value used to module the next token probabilities.
    param max_token: num of tokens to generate
    param return_probability: if set to True, return list of (samples, probability, tokens, logits)
    '''
    ret = []
    ret_prob = []
    ret_tokens = []
    ret_logits = []
    if top_p!=None and top_k!=None:
        for i in range(num_of_sentences):
            result = infer(prompt, top_p=top_p, top_k=top_k,  temp=temperature, gen_len=max_token,return_probability=return_probability)
            ret.append(result[0])
            if return_probability:
                ret_prob.append(result[1])
                ret_tokens.append(result[2])
                ret_logits.append(result[3])
    else:
        for i in range(num_of_sentences):
            result = infer(prompt, top_p=top_p, top_k=top_k, temp=temperature, gen_len=max_token, return_probability=return_probability)
            ret.append(result[0])
            if return_probability:
                ret_prob.append(result[1])
    if return_probability:
        return ret, ret_prob, ret_tokens,ret_logits
    return ret
```

## Neural Planner



### Workflow:

* nodes are in the form of **(character, action, precondition, solution)**
* Given: an action as a sentence s: 

* Create an empty node N0 (XX, s, XX, XX). add it to the empty graph G: 

* Put N0 in Q, a priority queue 

* While Q is not empty: 
>* N = pop Q 

>* Filling Slots & Use GPT to make precondition questions & Make a list of candidate preconditions base on N
>>* fill the character slot (c) by prompting GPT-J
>>* Perform syntax analysis on s.
>>* (optional) Use few-shot learning to teach GPT-J to expand the syntax and include keywords if possible. 
>>* if a keyword shows up in the expanded syntax, it means we need to ask certain type of questions.
>>>* For instance, "John was arrested"->"somebody was arrested"->"somebody was arrested `because` somebody did something"-> 'because' detected -> ask "why John was arrested?"
>>* Answer selected questions with GPT-J to fill in preconditions
>>* We have 4 types of questions and thus 4 types of preconditions. 
>>> `item needed`, `interaction with person`, `location`, `reason`
>* For each character preconditions pair (c,q). Determine if it is an open precondition
>>* If (c,q) is satified by default, ignore it.
>>* If (c,q) has already appeared somewhere in the path to N, discard node N, gather all unsolved preconditions caused by the deletion of N and add those unsolved preconditions to open conflict list.
>>* If (c,q) has already appeared somewhere but not in the path to N. Merge two branchs to reuse actions. Check forward to detect and remove all potential conflicts. 
>>* otherwise, append (c,q) to open-conflict slot. 

>* For each open precondition (c,p)
>>* generate a solution a with GPT-J
>>>* we only generate solutions for preconditions of type `item needed`, and `location`. We use the precondition directly as solutions for those of type `interaction with person`, and `reason`.
>>>* We buffer the generated solution so that we can reuse them if a solution is rejected due to confliction
>>>* We only generate solutions if we run out of buffered solutions.
>>* Duplicate node N. Add a new node N'=(character, s, p, a) to G
>>* Create a new node (XX, a, XX, XX). Push it to Q

```python
#from sys import last_traceback
beginning = "Fred is a normal American."
end = "Fred became rich"

import spacy
import copy
NLP = spacy.load("en_core_web_sm")


lemmatizer = nltk.stem.WordNetLemmatizer()

accessibility_prompt = '''Given the person's identity, judge if the statement about the person is true or false.

identity: normal American
statement: A normal American has a gun
Answer: No

identity: policeman
statement: A policeman has a gun
Answer: Yes

identity: normal American
statement: A normal American is homosexual
Answer: No

identity: normal American
statement: A normal American is evil
Answer: No

identity: {$text1$}
statement: A {$text1$} {$text2$}
Answer:'''

flaut_prompt = '''This is a precondition quiz. Determine what prerequisites need to be met to make the statement true.

statement: {$text1$} buys a hamburger
precondition: be at a KFC

statement: {$text1$} drives home
precondition: have a car

statement: {$text1$} deposits some money
precondition: be at a bank

statement: {$text1$} had got divorced with his wife
precondition: his wife cheat on him

statement: {$text1$} gets an A in his math course
precondition: studies hard

statement: {$text1$} {$text2$}
precondition:'''


satisfiability_check_prompt = '''Determine whether the statement is satisfied or not given the precondition only. 

preconditions: be at a KFC
statement: {$text1$} buy a hamburger
Answer: No

precondition: have a car
statement: {$text1$} drive home
Answer: Yes
Other preconditions: None

precondition: be at a bank
statement: {$text1$} deposit some money
Answer: No

precondition: his wife cheat on him
statement: {$text1$} had got divorced with his wife
Answer: Yes

precondition: studies hard
statement: {$text1$} get an A in his math course
Answer: Yes

preconditions: has money
statement: {$text1$} buy a book
Answer: No

preconditions: {$text2$}
statement: {$text1$} {$text3$}
Answer:'''



parallel_precondition_prompt = '''What preconditions do we need to satisfy to make a person buy a hamburger?
preconditions: the person has money; the person is at a restaurant; the restaurant has hamburger for sale

What preconditions do we need to satisfy to make a person get good grade in an exam?
preconditions: the person study hard; the exam is easy;

What preconditions do we need to satisfy to make a person {$text3$}?
preconditions:{$text2$}'''


parallel_precondition_item_prompt = '''Find out what item is required for the given circumstance, if nothing is needed answer nothing.

What item does {$text4$} need to walk to the bus station?
Answer: {$text4$} needs to have nothing,

What items does {$text4$} need to drive home to get to home?
Answer: {$text4$} needs to have car,

What item does {$text4$} need to meet someone?
Answer: {$text4$} needs to have nothing,

What item does {$text4$} need to watch a movie at the cinema?
Answer: {$text4$} needs to have a movie ticket,

What item does {$text4$} need to buy milk from the grocery store?
Answer: {$text4$} needs to have money,

What item does {$text4$} need to {$text3$}?
Answer: {$text4$} needs to have{$text2$}'''


parallel_precondition_item_prompt = '''Find out what item is required for the given circumstance, if nothing is needed answer nothing.

What item must {$text4$} possess to clean table?
Answer: {$text4$} must have a table cloth,

What item must {$text4$} possess to a man with a handgun shot {$text4$} on the street?
Answer: {$text4$} must have nothing,

What item must {$text4$} possess to rode his bike with friends?
Answer: {$text4$} must have a bike,

What item must {$text4$} possess to rob a bank?
Answer: {$text4$} must have a gun,

What item must {$text4$} possess to buy book from the book store?
Answer: {$text4$} must have money,

What item must {$text4$} possess to walk to the bus station?
Answer: {$text4$} must have nothing,

What item must {$text4$} possess to gave his parent a birthday gift?
Answer: {$text4$} must have a birthday gift,

What item must {$text4$} possess to shoot someone?
Answer: {$text4$} must have a gun,

What item must {$text4$} possess to watch a movie at the cinema?
Answer: {$text4$} must have a movie ticket,

What item must {$text4$} possess to run away from a bear?
Answer: {$text4$} musto have nothing,

What item must {$text4$} possess to take taxi?
Answer: {$text4$} must have money,

What item must {$text4$} possess to was sitting on a beach?
Answer: {$text4$} must have sunglasses,

What item must {$text4$} possess to got a gift from friends?
Answer: {$text4$} must have nothing,

What item must {$text4$} possess to buy milk from the grocery store?
Answer: {$text4$} must have money,

What item must {$text4$} possess to {$text3$}?
Answer: {$text4$} must have{$text2$}'''

veto_item_prompt = '''Determine whether the following statement is correct or not.

{$text4$} need to have tool, transportation method, currency, or ingredient to drive home.
Answer: True

{$text4$} need to have tool, transportation method, currency, or ingredient to walk home.
Answer: False

{$text4$} need to have tool, transportation method, currency, or ingredient to take a phone from table.
Answer: False

{$text4$} need to have tool, transportation method, currency, or ingredient to make pasta.
Answer: True

{$text4$} need to have tool, transportation method, currency, or ingredient to{$text3$}?
Answer:'''



parallel_precondition_action_prompt = '''How to buy something?
Answer: by ordering online;

How to get a good grade in an exam?
Answer: by studying hard

How to go to home?
Answer: by taking a bus;

How to {$text3$}?
Answer: by{$text2$}'''

parallel_precondition_second_character_prompt = '''What happens between the protagonist and another character prior to the event?
Event: Tom bought a second-hand car.
prior to Tom bought a second-hand car, Tom negotiated with a used car dealer. The car dealer convinced Tom to buy his car;

Event: Tim chats with Smith at a bar.
prior to Tim chats with Smith at a bar, Tim met Smith at the bar. Smith met Tim at the bar;

Event: Danie had an fight with Tony.
prior to Danie had an fight with Tony, Danie had an arguement with Tony. Tony had an arguement with Danie;

Event: Fred hugs Tom on street.
prior to Fred hugs Tom, Fred met Tom on street. Tom met Smith on street;

Event: {$text3$}.
prior to {$text3$},'''


parallel_precondition_second_character_prompt = '''What happens between the protagonist and another character prior to the event?
Event: Tom bought a second-hand car.
prior to Tom bought a second-hand car, Tom negotiated with a used car dealer;

Event: Tim chats with Smith at a bar.
prior to Tim chats with Smith at a bar, Tim met Smith at the bar;

Event: Danie had an fight with Tony.
prior to Danie had an fight with Tony, Danie had an arguement with Tony;

Event: Fred hugs Tom on street.
prior to Fred hugs Tom, Fred met Tom on street;

Event: {$text3$}.
prior to {$text3$},'''


parallel_precondition_location_prompt = '''Where does {$text2$} need to be to get food from the fridge?
Answer: {$text2$} needs to be at {$text2$} kitchen;

Where does {$text2$} need to be to watch a movie at the cinema?
Answer: {$text2$} needs to be at the cinema;

Where does {$text2$} need to be to take a suit from the wardrobe?
Answer: {$text2$} needs to be at {$text2$}'s home;

Where does {$text2$} need to be to take an airplane?
Answer: {$text2$} needs to be at airport;

Where does {$text2$} need to be to {$text3$}?
Answer: {$text2$} needs to be at'''

parallel_precondition_location_prompt = '''{$text2$} watch a movie.
Where is {$text2$}? cinema;

{$text2$} take a suit from the wardrobe.
Where is {$text2$}? {$text2$}'s home;

{$text2$} bought a pen.
Where is {$text2$}? store;

{$text2$} meet Fred at museum.
Where is {$text2$}? museum;

{$text2$} take an airplane.
Where is {$text2$}? airport;

{$text2$} see a bear.
Where is {$text2$}? forest;

{$text2$} {$text3$}.
Where is {$text2$}?'''

parallel_precondition_location_prompt = '''{$text2$} watch a movie.
Where does {$text2$} need to be to watch a movie? cinema;

{$text2$} take a suit from the wardrobe.
Where does {$text2$} need to be to take a suit from the wardrobe? home;

{$text2$} bought a pen.
Where does {$text2$} need to be to bought a pen? store;

{$text2$} meet Fred at museum.
Where does {$text2$} need to be to meet Fred at museum? museum;

{$text2$} take an airplane.
Where does {$text2$} need to be to take an airplane? airport;

{$text2$} see a bear.
Where does {$text2$} need to be to see a bear? forest;

{$text2$} {$text3$}.
Where does {$text2$} need to be to {$text3$}?'''

parallel_effect_prompt = '''how does a person have some food to eat food?
Answer: go to a restaurant; order takeout online；

how does a person study to get good grade in an exam?
Answer: revise to prepare for the exam;

how does a person by taking bus to go to the museum?
Answer: take bus;

how does a person at home to party with friends?
Answer: drive to home; walk home;

how does a person {$text3$}?
Answer:{$text2$}'''


action_prompt_obtain_item = '''Context: {$text2$} eats a hamburger.
How did {$text2$} get a hamburger?
Answer: {$text2$} bought a hamburger at a restaurant;

Context: {$text2$} opened the envelope.
How did {$text2$} get a knife?
Answer: {$text2$} gets a knife from the drawer;

Context: {$text2$} takes a shower.
How did {$text2$} get a towel?
Answer: {$text2$} gets a towel from towel rack;

Context: {$text2$} went to a concert.
How did {$text2$} get a ticket?
Answer: {$text2$} buys a ticket from the ticket office;

Context: {$text1$}
How did {$text2$} {$text3$}
Answer: {$text2$}'''

action_prompt_go_to_somewhere = '''Context: {$text2$} eats a hamburger at restaurant.
How did {$text2$} get to restaurant?
Answer: {$text2$} drove to get to the restaurant;

Context: {$text2$} had a stroke.
How did {$text2$} get to hospital?
Answer: {$text2$} was taken to hospital by ambulance;

Context: {$text2$} had a party at home.
How did {$text2$} get home?
Answer: {$text2$} walked home;

Context: {$text2$} greets his teacher at the classroom.
How did {$text2$} get to school?
Answer: {$text2$} took a bus to get to school;

Context: {$text1$}
How did {$text2$} {$text3$}
Answer: {$text2$}'''

parallel_reason_prompt = '''Explain what the protagonist has done that causes the event to happen.

Example 1: John went to hospital because Mike wants to see a doctor.

Example 2: Fred got zero in the exam because Fred cheated during the exam.

Example 3: Tom read a book because Tom wants to read a book.

Example 4: Tina was fired by her boss because Tina failed to finish her work.

Example 5: {$text3$} because{$text2$}'''

enrich_context_prompt='''Elaborate the given event to include a location where the event take place.

Original sentence: Tony watched a movie.
Sentence with a location where the event takes place: Tony watched a movie at the cinema.

Original sentence: Tina bought a pancake at supermarket.
Sentence with a location where the event takes place: Tina bought a pancake from a supermarket.

Original sentence: Mike went hiking.
Sentence with a location where the event takes place: Mike went hiking with his friends at a national park.

Original sentence: Jam was seen drinking at a bar by the bartender.
Sentence with a location where the event takes place: Jam was seen drinking at a bar by the bartender.

Original sentence: {$text3$}
Sentence with a location where the event takes place: {$text2$}'''


extract_character_name_prompt='''Name the principal character in the sentence. Here are some examples:
Example 1:
Sentence: "Hi, Tim", Mark said to Tim.
Characters: Mark,

Example 2: 
Sentence: Sherry went to bed.
Characters: Sherry,

Example 3:
Sentence: Jack met with Sam.
Characters: Jack

Example 4:
Sentence: the lawyer had an arguement with his wife.
Characters: the lawyer

Example 5:
Sentence: {$text3$}
Characters:'''

object_state_to_action_prompt = '''Given the sentence, what we tell about the state of the item in the sentence?

Context: Tim picked up a bottle of water from the table.
The bottle need to be full.
Tim take water from the tap.

Sentence: Millian shot Mike.
The gun needs to be loaded.
Millian loaded the gun.

Sentence: Jonny took a bus to get to home.
The bus ticket needs to be validated.
Jonny let the ticket collector to validate the ticket.

Sentence: Fred made steak.
The steak needs to be marinated.
Fred marinated the steak.

Sentence: Willson drove his car to Sally's home.
The car needs to be repaired.
Willson repair his car.

Sentence: Tony wrote with a typewriter.
The typewriter needs to be turned on.
Tony turned on the typewriter.

Sentence: {$text3$}
{$text4$}
{$text2$}'''

object_state_prompt = '''Given the sentence, what we tell about the state of the item in the sentence?

Sentence: Tim picked up a bottle of water from the table.
What can we tell about the bottle of water other than how Tim obtain it?
nothing.

Sentence: Millian shot Mike.
What can we tell about the gun other than how Millian obtain it?
The gun needs to be loaded.

Sentence: Jonny took a bus to get home.
What can we tell about the bus ticket other than how Jonny obtain it?
nothing.

Sentence: Fred made steak.
What can we tell about the steak other than how Fred obtain it?
The steak needs to be marinated.

Sentence: Willson drove his car to Sally's home.
What can we tell about the car other than how Willson obtain it?
nothing.

Sentence: Tony wrote with a typewriter.
What can we tell about the typewriter other than how Tony obtain it?
The typewriter needs to be turned on.

Sentence: Tom paid his rent.
What can we tell about the money other than how Tom obtain it?
nothing.

Sentence: {$text3$}
What can we tell about the {$text4$} other than how {$text2$} obtain it?
'''

from collections import defaultdict
text = ""
knowledge_graph = []
reasoning_graph = []
banned_preconditions_map = defaultdict(list)
ends = []
open_conflict = []
reusable_solutions = defaultdict(list)
context_hint = '''Here is some context for the event that's going to happen.
{$text1$}

'''

3
def get_prompt(prompt,text1="",text2="", text3="", text4="", text5=""):
    '''
    generate prompt with templates
    '''
    if "{$text1$}" in prompt:
        prompt = prompt.replace("{$text1$}",text1)
    if "{$text2$}" in prompt:
        prompt = prompt.replace("{$text2$}",text2)
    if "{$text3$}" in prompt:
        prompt = prompt.replace("{$text3$}",text3)
    if "{$text4$}" in prompt:
        prompt = prompt.replace("{$text4$}",text4)
    if "{$text5$}" in prompt:
        prompt = prompt.replace("{$text5$}",text5)
    return prompt


def preprocess(candidate_preconditions):
  '''
  preprocess before generating actions for preconditions
  '''
  if isinstance(candidate_preconditions,str):
    if "has done" in candidate_preconditions:
      candidate_preconditions = candidate_preconditions.replace("has done ","")
    if "has " in candidate_preconditions:
      candidate_preconditions = candidate_preconditions.replace("has ","get ")
    if "at " in candidate_preconditions:
      candidate_preconditions = candidate_preconditions.replace("at ","get to ")
    return candidate_preconditions
  else:
    ret = []
    for precondition in candidate_preconditions:
      if "has done" in precondition:
        precondition = precondition.replace("has done ","")
      if "has " in precondition:
        precondition = precondition.replace("has ","get ")
      if "at " in precondition:
        precondition = precondition.replace("at ","get to ")
      ret.append(precondition)
    return ret

def lemmarization(word):
    return lemmatizer.lemmatize(word)

def match_lemma(precondition1,precondition2,threshold=0.6):
    word_lemma1 = [lemmarization(word) for word in precondition1.split(' ')]
    word_lemma2 = [lemmarization(word) for word in precondition2.split(' ')]
    
    u=len(set(word_lemma1).union(set(word_lemma2)))
    i = len(word_lemma1)+len(word_lemma2)-u
    if i/u>threshold:
        return True
    else:
        return False

def get_verb(sentence):
  doc = NLP(sentence)
  for token in doc:
    if token.pos_ == 'VERB':
      return token.text, token.lemma_
  return "",""

def choose_best_action_by_perplexity(effect,actions):
    true_false_prompt = '''Determine if the statement is true or false
    statement: {$effect$} is the result of {$action$} 
    answer: true'''
    scores = [score(true_false_prompt.replace("{$action$}",action).replace("{$effect$}",effect)) for action in actions]
    print(scores)
    return actions[scores.index(min(scores))]


def choose_best_precondition_by_perplexity(action,preconditions):
    '''
    return the best precondition for certain action according to perplexity
    param action: action
    param preconditions: candidate preconditions for the action
    '''
    true_false_prompt = '''Determine if the statement is true or false
    statement: One need to {$precondition$} to {$action$}
    answer: true'''
    scores = [score(true_false_prompt.replace("{$precondition$}",precondition).replace("{$action$}",action)) for precondition in preconditions]
    print(scores)
    return preconditions[scores.index(min(scores))]


def choose_precondition_by_frequency(preconditions, freq_threshold=2,cutoff_frequency=0,action="",match_lemma_threshold=0.66, use_majority_vote = True):
    '''
    choose most common preconditions from candidates, a candidate precondition is considered as valid once its frequency pass this threshold
    When none of candidate preconditions pass the threshold, the most frequent one will be returned if its frequency exeed cutoff_frequency.
    when action is given, return the best suit precondition given the action.
    param preconditions: candidate preconditions
    param freq_threshold: a candidate precondition is considered as valid once its frequency pass this threshold
    param cutoff_frequency: When none of candidate preconditions pass the threshold, the most frequent one will be returned if its frequency exeed cutoff_frequency.
    param action: when action is given, return the best suit precondition given the action.
    param match_lemma_threshold: between 0 and 1, a higher value indicate less tolerance in fuzzy matching
    param use_majority_vote: if True, return the majority vote result (More than 50% of votes)
    '''
    pool = {}
    for precondition in preconditions:
        pool[precondition] = 0
    for precondition in preconditions:
        for key in pool.keys():
            if match_lemma(precondition,key,threshold=match_lemma_threshold):
                pool[key]+=1
    most_common_freq = 0
    most_common_precondition = None
    ret = []
    pool = dict(sorted(pool.items(),key=lambda x:x[1],reverse=True))
    print(pool)
    for key in pool.keys():  
        if use_majority_vote and pool[key]>=len(preconditions)/2.0:
            return [key]  
        if pool[key]>=most_common_freq:
            most_common_precondition = key
            most_common_freq = pool[key]
        if pool[key]>=freq_threshold:
            ret.append(key)
    if most_common_precondition not in ret and most_common_freq>cutoff_frequency:
        ret.append(most_common_precondition)
    if action!="" and ret!=[]:
        ret=[choose_best_precondition_by_perplexity(action, ret)]
    return ret

def post_process(generated_sentences,separator=';',prefix='', mode="gen_preconditions", must_have='',must_not_contain = '', to_avoid = "",action="",remove_first_word=False,exclude_go_to=False,
                 lookup={},simi_threshold=0.8,freq_threshold=20,cutoff_frequency=0,match_lemma_threshold=0.66, use_majority_vote = True):
    '''
    all-in-one postprocess function for postprocessing outputs of GPT-J.
    param generated_sentences: original GPT-J output
    param seperator: a string for the delimiter
    param prefix: added as the prefix to the postprocess results
    param mode: string, if "gen_actions", filter all candidates that are similar to to_avoid
    param must_have: string, filter out candidates that doesn't contain it
    param to_avoid: string, must be non-empty if mode is "gen_actions"
    param action: for choose_best_precondition_by_perplexity
    remove_first_word: whether or not to remove the first word, usually the name of the protagonist
    param lookup: not used
    '''
    candidates = []
    #print(generated_sentences)
    for result in generated_sentences:
        result = result.split('\n')[0]
        if separator in result:
            result = [(s.strip()).strip() for s in result.split(separator)[:-1]]
        else:
            result = [(s.strip()).strip() for s in result.split(separator)]
        if mode=="gen_actions":
          for action_ in result:
            similarity_score = cos_simi(to_avoid,action_.split('.')[0])
            if similarity_score<simi_threshold and len(action_.replace('.','').strip().split(' '))>1:
                if ' is ' not in action_ and ' was ' not in action_:
                    candidates.append(action_)
        else:
          candidates.extend(result)
    candidates = [candidate.strip() for candidate in candidates]
    candidates = [candidate for candidate in candidates if len(candidate)>0]   #get rid of empty string
    if must_have!='':
        candidates = [candidate for candidate in candidates if must_have in candidate]
    if must_not_contain!='':
        candidates = [candidate for candidate in candidates if must_not_contain not in candidate]
    if len(candidates)==0:
        return candidates
    print("candidates",candidates)
    for i in range(len(candidates)):
        if candidates[i].split(' ')[0] in ['a','an','the'] or remove_first_word:
            candidates[i] = ' '.join(candidates[i].split(' ')[1:])
    if exclude_go_to:
        candidates = [candidate for candidate in candidates if 'go to' not in candidate and 'goes to' not in candidate]
    candidates = choose_precondition_by_frequency(candidates, freq_threshold=freq_threshold, cutoff_frequency =cutoff_frequency, 
                                                  action=action, match_lemma_threshold=match_lemma_threshold, use_majority_vote=use_majority_vote)
    if mode=="gen_item":
        candidates = [candidate for candidate in candidates if candidate not in ['nothing','something']]
    return [prefix+' '+candidate for candidate in candidates]

def fix_missing_object(sentence,item):
    '''
    replace any pronom with specific object
    '''
    tokens = nltk.word_tokenize(sentence)
    tagged = nltk.pos_tag(tokens)
    ret = ''
    for word in tagged:
        new_word = word[0]
        if word[1]=='PRP':
            new_word = item
        ret+=new_word+' '
    return ret.strip()

def backtrack(target,character):
    '''
    backtrack, return the logic chain (context) that explains why a target event happen
    param target: target event
    already_contains_item_state: the number of item state contraints in all sub causal chains, used to terminate recursion.
    '''
    global banned_preconditions_map
    #print(target,character,'.')
    ids = []
    new_targets = []
    logic_chain = ""  #"To John was arrested. (John was arrested at his house. because  John's wife saw him taking drugs) , the person need to has police, which is solved by was arrested at his house."
    visited = []
    already_contains_item_state = 0
    target_ = target.replace('.','')
    banned_preconditions_map[(target_,character)] = []
    if (target_,character) not in banned_preconditions_map[(target_,character)]:
        banned_preconditions_map[(target_,character)].append((target_,character))
    preconditions = []
    for entry in knowledge_graph:
        if entry[2]==target and entry[3]==character:
            preconditions.append(entry[1])
    preconditions = list(set(preconditions))
    explored = [target_,target]
    for entry in knowledge_graph:
        if entry[1] in preconditions and entry[0][0] not in explored and entry[3]==character:
            explored.append(entry[0][0])
            logic_chain_next, pending_number_item_state = backtrack(entry[0][0],find_character_names(entry[0][0])[0])
            already_contains_item_state += pending_number_item_state
            logic_chain+=logic_chain_next
            banned_preconditions_map[(target_,character)].extend(banned_preconditions_map[(entry[0][0].replace('.',''),entry[3])])
            banned_preconditions_map[(target_,character)].append((entry[1][0],entry[3]))
            banned_preconditions_map[(target_,character)] = list(set(banned_preconditions_map[(target_,character)]))
            if 'which is satisfied automatically.'!=entry[2] and entry[2]==target:
                logic_chain+='\nTo '+entry[0][0]+' ('+entry[0][1]+'), '+entry[3]+' need to '+entry[1][0]+', which is solved by '+entry[2]+'.'
                if entry[1][1]=='item state':
                    already_contains_item_state += 1

    return logic_chain, already_contains_item_state

    
def forward_traverse(end):
    '''
    Used for coverting a DAG to a plan in text
    param end: end node of the plan (end[0] is the input to the pipeline)
    '''
    edges = [end]
    ids = []
    while len(edges)>0:
        #print(edges)
        edge = edges.pop(0)
        for id,node in enumerate(knowledge_graph):
            if id not in ids and edge==node[0][0]:
                edges.append(node[3]+' '+node[2] if node[3]!=node[2].split(' ')[0].strip() else node[2])
                ids.append(id)
    formated_output = '\n'.join(['To '+knowledge_graph[id][0][0]+' ('+knowledge_graph[id][0][1]+'), '+knowledge_graph[id][3]+' need to '+knowledge_graph[id][1][0]+', which is solved by '+knowledge_graph[id][2] for id in ids])
    return formated_output  

def check_contains_upper(s):
    for c in s:
        if c.isalpha() and c == c.upper():
            return True
    return False 

def find_character_names(sentence):
    '''
    return the principal character's name
    '''
    extract_principle_character_name_prompt='''Name the principal character in the sentence. Here are some examples:
Example 1:
Sentence: "Hi, Tim", Mark said to Tim.
Characters: Mark,

Example 2: 
Sentence: Sherry went to bed.
Characters: Sherry,

Example 3:
Sentence: Jack met with Sam.
Characters: Jack

Example 4:
Sentence: the lawyer had an arguement with his wife.
Characters: the lawyer

Example 5:
Sentence: {$text3$}
Characters:'''
    extract_all_characters_name_prompt='''Name all characters in the sentence. Here are some examples:
Example 1:
Sentence: "Hi, Tim", Mark said to Tim.
Characters: Mark, Tim,

Example 2: 
Sentence: Sherry went to bed.
Characters: Sherry,

Example 3:
Sentence: Jack met with Sam.
Characters: Jack, Sam,

Example 4:
Sentence: the lawyer had an arguement with his wife.
Characters: the lawyer, the lawyer's wife,

Example 5:
Sentence: Mike works hard at Apple Inc.
Characters: Mike,

Example 6:
Sentence: {$text3$}
Characters:'''
    prompt = get_prompt(extract_principle_character_name_prompt,"",'',sentence)
    output = gpt_call(prompt, 6, 1, top_p=0.5, top_k=1)
    #print(output)
    major_characters = post_process(output,prefix="",mode="gen_item",freq_threshold=0, cutoff_frequency=0,  match_lemma_threshold=0.9,separator=',')

    prompt = get_prompt(extract_all_characters_name_prompt,"",'',sentence)
    output = gpt_call(prompt, 6, 1, top_p=0.5, top_k=1)
    #print(output)
    all_characters = post_process(output,prefix="",mode="gen_item",freq_threshold=0, cutoff_frequency=0,  match_lemma_threshold=0.9,separator=',',use_majority_vote=False)
    all_characters = [ch.strip() for ch in all_characters if check_contains_upper(ch)]
    if len(major_characters)==0:
        return "PersonX"
    else:
        return major_characters[0].strip(),all_characters
            

def forward_fix_banned_preconditions(merge_point,banned_preconditions):
    '''
    forward check for cycle when a new branch is merged to an existing branch.
    update banned_precondition along its path. If a cycle is detect (a precondition that is banned show up in the existing branch), 
    break the cycle by discarding everything after the actions that requires that precondition (inclusively) and mark preconditions 
    of those actions on the edge as opening conflict
    param merge_point: a precondition where two branchs merge
    param banned_preconditions: banned_preconditions given by the new branch
    '''
    global banned_preconditions_map, knowledge_graph, ends, open_conflict
    ids = []
    for id,node in enumerate(knowledge_graph):
        if id not in ids and merge_point[0]==node[1][0] and merge_point[1]==node[3]:
            ids.append(id)   #actions that require the first common precondition of the two branchs
    pending = []
    for id in ids:
        action = knowledge_graph[id][2].replace('.','')   
        character = knowledge_graph[id][3]
        banned_preconditions_map[(action,character)].extend(banned_preconditions)
        banned_preconditions_map[(action,character)] = list(set(banned_preconditions_map[(action,character)]))
        for node in knowledge_graph:
            if node[3]!=character:
                continue
            if action==node[0][0].replace('.',''):
                precondition = (node[1][0],node[3])
                if precondition in banned_preconditions_map[(action,character)]:
                    for idx,entry in enumerate(knowledge_graph):
                        if entry[2].replace('.','')==action:
                            print("----forward_fix_banned_preconditions: roll back to:",[copy.deepcopy(entry[0]),copy.deepcopy(entry[1])])
                            open_conflict.append((copy.deepcopy(entry[0][0]),copy.deepcopy(entry[0][1]),copy.deepcopy(entry[1][0]),copy.deepcopy(entry[1][1])),copy.deepcopy(entry[3]))
                            knowledge_graph[idx][0] = ("Dummy Sink","Dummy Sink")
                            knowledge_graph[idx][1] = ("Dummy Sink","Dummy Sink")
                            knowledge_graph[idx][2] = "XXXXXXXXXX"
                else:
                    pending.append((precondition,list(set(banned_preconditions_map[(action,node[3])]+[precondition]))))
    for next_precondition,next_banned_precondition in pending:
        #print(next_precondition)
        forward_fix_banned_preconditions(next_precondition,next_banned_precondition)

def calculate_PMI_dc(character, action, hypothesis, pmi_prompt_full, pmi_prompt_domain,question_type='multiQA'):
    '''
    return the PMI support for each hypothesis by formulating it as a multi-choice problem.
    Important Note: PMI trick only works well when the probability of each hypothesis is non-negligible
    param character: string character name
    param action: the action for which certain preconditions need to be satisfied
    param hypothesis: preconditions being ranked
    param pmi_prompt_full: full prompt, contains question, choices, (examples). Read this paper "Surface Form Competition: Why the Highest Probability Answer Isn’t Always Right" for more detail
    param pmi_prompt_domain: usually just "?\nAnswer:"
    '''
    print(action)
    cnt = 0
    PMI_dc = {}
    d = {}    
    pmi_prompt_full_ = str(pmi_prompt_full)
    hypothesis = hypothesis[:8]
    if question_type=='multiQA':
        token_map = {'A':317,'B':347,'C':327,'D':360,'E':412,'F':376,'G':402,'H':367,'I':314}
        option_template = ['{$optionA$}','{$optionB$}','{$optionC$}','{$optionD$}','{$optionE$}','{$optionF$}','{$optionG$}','{$optionH$}','{$optionI$}']
        tokens = list(token_map.keys())
        for i in range(len(hypothesis)):
            PMI_dc[hypothesis[i]]=0
            d[tokens[i]] = hypothesis[i]
            pmi_prompt_full_ = pmi_prompt_full_.replace('{$option'+tokens[i]+'$}', '\n'+tokens[i]+':'+hypothesis[i])
    elif question_type=='TrueFalse':
        token_map = {'Yes':3363,'No':1400,'yes':3763,'no':645}
        option_template = []
        PMI_dc['Yes']=0
        PMI_dc['No']=0
        d['Yes'] = 'Yes'
        d['No'] = 'No'
        d['yes']='Yes'
        d['no']='No'

    #print(d)
    #print(pmi_prompt_full_)
    for template in option_template:
        pmi_prompt_full_ = pmi_prompt_full_.replace(template, '')
    if question_type=='multiQA':
        pmi_prompt_full_ = get_prompt(pmi_prompt_full_, '','',action,character)
    elif question_type=='TrueFalse':
        pmi_prompt_full_ = get_prompt(pmi_prompt_full_, '',' '.join(hypothesis[0].split(' ')[1:]),action,character)    
    #print(pmi_prompt_full_)
    pmi_prompt_domain_ = get_prompt(pmi_prompt_domain)
    _, _, _, logits_full = gpt_call(pmi_prompt_full_, 3, 1, top_p=0.8, top_k=100, return_probability=True)
    _, _, _, logits_domain = gpt_call(pmi_prompt_domain_, 3, 1, top_p=0.8, top_k=100, return_probability=True)
    for k in d.keys():
        PMI_dc[d[k]]+=logits_full[0][0][token_map[k]]/logits_domain[0][0][token_map[k]]
    return PMI_dc


def get_double_negation(sentence, num_generations = 1, verbose = False):
    neg_output = negate_input(sentence, num_generations = num_generations)

    if verbose:
      print("input negation: ", neg_output)

    cfp_gens = counter_fact_precond(neg_output[0][:-1], num_generations = num_generations)

    uniq_cfp_gens, counts = np_unique_sent(cfp_gens, return_counts = True)
    uniq_cfp_gens, counts = list_sort(uniq_cfp_gens, counts)
    uniq_cfp_gens.reverse()
    counts.reverse()

    if verbose:
      print("cfpreconditions: ", uniq_cfp_gens)
      print("counts: ", counts)

    negations = []

    for gen in uniq_cfp_gens:
      all_negations, neg_counts = np_unique_sent(negate_input(gen), return_counts = True)
      all_negations, neg_counts = list_sort(all_negations, neg_counts)

      if verbose:
        print("all candidate negations for ", gen)
        print(all_negations)
        print(neg_counts)

      negations.append(all_negations[np.argmax(neg_counts)])
    
    return negations

def negate_input(sentence, num_generations = 5):
  '''
  return the negation of the input sentence
  param sentence: string. the action or statement
  param generations: int. the number of times to query GPT
  '''

  negation_prompt_step1 = '''Hanna got in the car. The negation of this is Hanna didn't get in the car.
  Steve went to school. The negation of this is Steve didn't go to school.
  Bob gave Jenny a key. The negation of this is Bob didn't give Jenny a key.
  Timmy didn't have a gun. The negation of this is Timmy had a gun.
  {$text1$} The negation of this is'''

  all_gens = []
  prompt = get_prompt(negation_prompt_step1, sentence)

  while len(all_gens) < num_generations:
    gpt_output = gpt_call(prompt, 10, 15, top_p=0.6, top_k=10, return_probability=False)
    output           = gpt_output[0].split("\n")[0]
    # print("output: ", output)
    for atom in split_sentence_array(output):
      if len(atom) > 1 and ("_" not in atom):
        all_gens.append(atom)
  
  return all_gens


def counter_fact_precond(sentence, num_generations = 5):
  '''
  Given a statement, expand on the cause or potential reason behind the statement.
  param sentence: string. the action or statement.
  param generations: int. the number of times to query GPT
  '''

  negation_prompt_step2 = '''Hanna didn't give Bob an apple because Bob was allergic to apples.
  Steve didn't go to the park. This was because the park was closed.
  Jen didn't go into her house. This was because Jen's house was locked.
  Ben didn't go to school. This was because a scary clown was at school.
  Ethan didn't drive to work. This was because Ethan's car was out of gas.
  Jake wasn't hungry. This was because Jake ate a burger.
  {$text1$}. This was because'''

  all_gens = []
  prompt = get_prompt(negation_prompt_step2, sentence)

  max_attempts = 10
  attempts     = 0

  while len(all_gens) < num_generations and attempts < max_attempts:
    gpt_output = gpt_call(prompt, 10, 15, top_p=0.6, top_k=10, return_probability=False)
    output           = gpt_output[0].split("\n")[0]
    # print("output: ", output)
    for atom in split_sentence_array(output):
      if len(atom) > 1 and ("_" not in atom):
        all_gens.append(atom)

    attempts +=1
    
  return all_gens

def list_sort(items, counts, return_reverse = False):
  '''
  Given a list of items and their counts, return the list sorted by counts
  param items: list. an array of unique items
  param counts: list. an array of integers
  '''
  counts, items = (list(t) for t in zip(*sorted(zip(counts, items))))
  
  if return_reverse:
    items.reverse()
    counts.reverse()
    return items, counts
  else:
    return items, counts

def np_unique_sent(array, simi_threshold = .9, return_counts = False):
  '''
  Given a list of statments, find the unique elements and their counts such that no element shares a cosine similarity above a certain threshhold
  param array: list. An array of non-unique sentences.
  param simi_threshold: float. The threshold for the cosine similarity.
  param return_counts: bool. A flag for whether or not to return the counts as well as the unique elements
  '''
  new_array = []
  counts    = []
  for candidate in array:
    cos_score     = 0
    cur_cos_score = 0
    cur_counts    = 1
    for i, element in enumerate(new_array):
      cur_cos_score = cos_simi(candidate, element)
      if cur_cos_score > simi_threshold:
        cos_score   = cur_cos_score
        counts[i]  += 1
        break
      if cos_score < cur_cos_score:
        cos_score   = cur_cos_score

    if cos_score < simi_threshold:
      new_array.append(candidate)
      counts.append(cur_counts)

  if return_counts:
    return new_array, counts
  else:
    return new_array

def get_negation(sentence, character):
    PMI_prompt_full_negation = "Why {$text4$} fails to{$text3$}?'{$optionA$}{$optionB$}{$optionC$}{$optionD$}{$optionE$}{$optionF$}{$optionG$}{$optionH$}{$optionI$}\nAnswer:"
    PMI_prompt_domain_negation = "Why {$text4$} fails to  ?{$optionA$}{$optionB$}{$optionC$}{$optionD$}{$optionE$}{$optionF$}{$optionG$}{$optionH$}{$optionI$}Answer:"
    candidate_negations = call_comet(head=sentence.replace('.','').strip(),character=character)
    candidate_len = len(candidate_negations)
    cnt = 0
    PMI_dc = {}
    d = {}
    hypothesis = candidate_negations[:8]
    token_map = {'A':317,'B':347,'C':327,'D':360,'E':412,'F':376,'G':402,'H':367,'I':314}
    pmi_prompt_full_ = str(PMI_prompt_full_negation)
    pmi_domain_ = str(PMI_prompt_domain_negation)
    for i in range(len(hypothesis)):
        d[chr(ord('A')+i)] = hypothesis[i]
        pmi_prompt_full_ = pmi_prompt_full_.replace('{$option'+chr(ord('A')+i)+'$}', '\n'+chr(ord('A')+i)+':'+hypothesis[i])
    print(d)
    for template in ['{$optionA$}','{$optionB$}','{$optionC$}','{$optionD$}','{$optionE$}','{$optionF$}','{$optionG$}','{$optionH$}','{$optionI$}']:
        pmi_prompt_full_ = pmi_prompt_full_.replace(template, '')
    for i in range(len(hypothesis)):
        pmi_domain_ = pmi_domain_.replace('{$option'+chr(ord('A')+i)+'$}', '\n'+chr(ord('A')+i))
    for template in ['{$optionA$}','{$optionB$}','{$optionC$}','{$optionD$}','{$optionE$}','{$optionF$}','{$optionG$}','{$optionH$}','{$optionI$}']:
        pmi_domain_ = pmi_domain_.replace(template, '')
    pmi_prompt_full_ = get_prompt(pmi_prompt_full_, '','',sentence.replace(character,''),character)
    print(pmi_prompt_full_)
    pmi_prompt_domain_ = get_prompt(pmi_domain_)
    _, _, _, logits_full = gpt_call(pmi_prompt_full_, 3, 1, top_p=0.8, top_k=100, return_probability=True)
    _, _, _, logits_domain = gpt_call(pmi_prompt_domain_, 3, 1, top_p=0.8, top_k=100, return_probability=True)
    for k in d.keys():
        PMI_dc[d[k]]=logits_full[0][0][token_map[k]]/logits_domain[0][0][token_map[k]]
    best = -1
    best_negation = ""
    for k in PMI_dc:
        if PMI_dc[k]>best:
            best = PMI_dc[k]
            best_negation = str(k)
    if len(best_negation)>0:
        return [best_negation]
    else:
        return []


def split_sentence_array(array, end_token = ".", verbose = False):
  '''
  Given a string of sentences, split the sentences into an array and return
  param array: string. The string of sentences to be split
  param end_token: char. the end of sentence token for the string
  '''
  if not array:
    return []
  return_array = array.split(end_token + " ")
  ra = []
  if verbose:
    print("split sentences: ", return_array)

  for element in return_array:
    if element:
      if element[-1] != end_token:
        ra.append(strip_whitespace(element + end_token))
      else:
        ra.append(strip_whitespace(element))

  return ra

def strip_whitespace(string):
  '''
  Given a string, remove all blank spaces from the start and end.
  param stri
  '''
  for i, letter in enumerate(string):
    if letter != " ":
      string = string[i:]
      break
      
  #Reverse string and get other end
  string = string[::-1]
  for i, letter in enumerate(string):
      if letter != " ":
        string = string[i:]
        break

  return string[::-1]

def get_facts_and_check_ownership(sentence, obj, location):
    '''
    return a name of a person who owns the object
    '''
    fact_about_object_prompt = '''Given the sentence, what can be said about the object in the sentence?
Sentence: Tim picked up a bottle of water from the table.
What can we tell about the bottle of water?
Answer: There is a bottle of water on the table

Sentence: Tim bought milk from the grocery store.
What can we tell about milk?
Answer: There is milk for sale at the grocery store.

Sentence: {$text3$}
What can we tell about the {$text4$}?
Answer: There is'''
    ownership_prompt = '''Given the sentence, determine Who owns the object.
Sentence: There is a bottle of water on the table at John's home.
Who owns the bottle of water?
Answer: John.

Sentence: There is milk for sale at the grocery store.
Who owns the milk?
Answer: grocery store.

Sentence:{$text3$}
Who owns the {$text4$}?
Answer:'''
    facts_generation = 'There is'+gpt_call(get_prompt(fact_about_object_prompt,"","",sentence,obj), 8, 1, top_p=0.6, top_k=4)[0].split('\n')[0]
    ownership = gpt_call(get_prompt(ownership_prompt,"","",facts_generation.split('.')[0]+' at '+location+'.',obj), 8, 1, top_p=0.6, top_k=4)[0].split('\n')[0].replace('.','').strip()
    return facts_generation,ownership

def get_item_slot_multiple_items(sentence, context='',character=""):
    '''
    fill the slot about what is needed for the character to perform the action. Currently, we restrict it to generate only the most needed item for robustness issue.
    We use few-shot learning to guide GPT to generate 'nothing' if no item is needed. The nothing flag is then removed and will return an empty list of needed items
    Important Note: PMI trick only works well when the probability of each hypothesis is non-negligible. Therefore, we need to filter out outliers first
    param sentence: the action
    param context: string, context
    param character: string
    '''
    check_prompt_needItem = '''{$text1$} needs {$text2$} to {$text3$}  True'''
    check_prompt_becauseNeeds = '''Determine if the following statement is true?\n{$text3$} because {$text1$} needs {$text2$}.  True'''

    PMI_prompt_full_item_QA = '''What tool, transportation method, currency, or ingredient does {$text4$} need to{$text3$}?{$optionA$}{$optionB$}{$optionC$}{$optionD$}{$optionE$}{$optionF$}{$optionG$}{$optionH$}{$optionI$}\nAnswer:'''
    PMI_prompt_domain_item_QA = '''What tool, transportation method, currency, or ingredient does  need to  ?\nA:\nB:\nC\nD\nAnswer:'''
    PMI_prompt_full_item_flip = '''money is required to buy milk.
Is this sentence logically correct?
Answer: Yes
milk is required to buy milk.
Is this sentence logically correct?
Answer: No
nothing is required to walk to the post office.
Is this sentence logically correct?
Answer: Yes
{$text2$} is required to {$text3$}.
Is this sentence logically correct?
Answer:'''
    PMI_prompt_domain_item_flip = '''money is required to buy milk.
Is this sentence logically correct?
Answer: Yes
milk is required to buy milk.
Is this sentence logically correct?
Answer: No
nothing is required to walk to the post office.
Is this sentence logically correct?
Answer: Yes
.
Is this sentence logically correct?
Answer:'''
    PMI_prompt_full_is_item = '''Sentence 1: money is an item that can be owned.
Is this statement correct?
Answer: Yes

Sentence 2: arguement is an item that can be owned.
Is this statement correct?
Answer: No

Sentence 3:{$text2$} is an item that can be owned.
Is this statement correct?
Answer:'''
    PMI_prompt_domain_is_item = '''Sentence 1: money is an item that can be owned.
Is this statement correct?
Answer: Yes

Sentence 2: arguement is an item that can be owned.
Is this statement correct?
Answer: No

Sentence 3:
Is this statement correct?
Answer:'''


    temp = []
    ret = []
    trial_times = [4,1,1]
    trial_top_p = [0.7,0.8,0.8]
    cutoff_frequencies = [0, 1, 1]
    freq_thresholds = [4,1,1]
    for item_cnt in range(3):
        for trial in range(trial_times[item_cnt]):
            text2 = ''
            if len(temp)>0:
                text2 = ''.join([" "+item+',' for item in temp])
            action = ' '.join(sentence.split(' ')[1:]).replace('.','')
            prompt = get_prompt(parallel_precondition_item_prompt,"",text2,action,character)
            #print(prompt)
            
            #filter out outliers first
            output = post_process(gpt_call(prompt, 3, 15, top_p=trial_top_p[item_cnt], top_k=10, return_probability=False),prefix="",freq_threshold=2,must_not_contain='no ',match_lemma_threshold=0.2, separator=',')
            output = [out for out in output if not any([k in out for k in [' of',' with',' as',' for', ' to']])]
            if ' nothing' in output:
                output = output[:max(3,len(output)//2)]
                if ' nothing' not in output:
                    output.append(' nothing')
            else:
                output = output[:max(3,len(output)//2)]
            output.reverse()
            PMI_dc = calculate_PMI_dc(character,' '+action,output,PMI_prompt_full_item_QA, PMI_prompt_domain_item_QA)
            '''
            PMI_dc = {}
            for candidate in output:
                PMI_dc[candidate] = calculate_PMI_dc(character,action,['has'+candidate],PMI_prompt_full_item_flip,PMI_prompt_domain_item_flip,question_type='TrueFalse')
                print('has'+candidate,PMI_dc[candidate])
                PMI_dc[candidate] = PMI_dc[candidate]['Yes']/PMI_dc[candidate]['No']
            '''
            print(PMI_dc)
            if len(PMI_dc)==0:
                return []
            output = list(PMI_dc.keys())
            probability = [PMI_dc[k] for k in output]
            ret.append(output[probability.index(max(probability))].strip())
            ret = ['has '+r for r in ret if ('nothing' not in r and 'something' not in r)]
            if ret==[]:
                return []
            PMI_dc_is_item = calculate_PMI_dc('','',ret,PMI_prompt_full_is_item, PMI_prompt_domain_is_item,question_type='TrueFalse')
            print(ret, PMI_dc_is_item)
            if PMI_dc_is_item["Yes"]<=PMI_dc_is_item["No"]:
                return []
            return ret
        if temp==[]:
            return ret
    return ret


def get_slots_gpt(sentence,context_,last_precondition_type, num_item_state_in_the_plan=0):
    '''
    filling slots (preconditions) given the action. We considered four type of preconditions "item needed", "reason"(why?), "location"(where?), "interaction with other person".
    param sentence: string. the action
    param context_: string. context that explains why the action is take
    '''
    global reasoning_graph
    sentence_ = str(sentence)
    character,characters = find_character_names(sentence)
    characters = [ch.strip() for ch in characters]
    print('*****debug context*****')
    print("Character:",character)
    print(sentence)
    print(context_)
    #context_ = context_hint.replace('{$text1$}',text)
    print('fill slot.. ',sentence)
    slot_flag = try_expand_syntax(get_syntax_prompting(sentence.replace('.','')))
    if len(sentence.split(' '))<4:
        sentence = ' '.join(sentence.split(' ')).replace('.','')+' '+gpt_call(context_+get_prompt(enrich_context_prompt,"",' '.join(sentence.split(' ')).replace('.',''),' '.join(sentence.split(' '))), 13, 1, top_p=0.9, top_k=40)[0].split('\n')[0].strip()
        #sentence = gpt_call(context_+get_prompt(enrich_context_prompt,"","",' '.join(sentence.split(' '))), 13, 1, temperature=0.9)[0].split('\n')[0].strip()
        print("Expand Sentence:",sentence)
    slot = {}
    slot['item needed'] = []
    slot['reason'] = []
    slot['interaction with other person'] = []
    slot['location'] = []
    if last_precondition_type=="interaction with other person" or last_precondition_type=='reason':
        slot_flag['negation']=False
    #slot_flag['negation']=False
    if slot_flag['negation']==True:
        slot['negation'] = get_negation(sentence_,character)
        slot['item needed'] = []
        slot['reason'] = []
        slot['interaction with other person'] = []
        slot['location'] = []
        return slot_flag, slot, sentence, character
    if last_precondition_type == 'location':
        slot_flag['location']=False
    print(slot_flag)
    if slot_flag['location']==True:
        #print(context_+get_prompt(parallel_precondition_location_prompt,"",character,' '.join(sentence.split(' ')[1:])))
        slot['location'] = []
        for trial in range(4):
            sentence = str(sentence_)
            for person in characters:
                location=post_process(gpt_call('\n'.join(context_.split('\n')[-3:])+get_prompt(parallel_precondition_location_prompt,"",person,' '.join(sentence.split(' ')[1:])), 3, 15, temperature=0.9),prefix="at",freq_threshold=5, use_majority_vote=False)
                #location=post_process(gpt_call(get_prompt(parallel_precondition_location_prompt,"",person,' '.join(sentence.split(' ')[1:])), 3, 15, temperature=0.9),prefix="at",freq_threshold=5, use_majority_vote=False)
                #print('\n'.join(context_.split('\n')[-5:])+get_prompt(parallel_precondition_location_prompt,"",person,' '.join(sentence.split(' ')[1:])),location)
                if location!=[]:
                    if ' at ' not in sentence and ' from ' not in sentence and person==character:
                        sentence = sentence.split('.')[0].split(' at ')[0]+' '+location[0]
                    if person not in [l[0] for l in slot['location']]:
                        slot['location'].append((person,location[0]))
            if len(slot['location'])==len(characters):
                break
    if slot_flag['item']==True:
        slot['item needed']=get_item_slot_multiple_items(sentence, '', character)
        slot['item needed'] = [(character,item) for item in slot['item needed'] if ' no ' not in item]
        if slot['item needed']!=[] and num_item_state_in_the_plan==0:
            #print(get_prompt(object_state_prompt,"",character,sentence,slot['item needed'][0][1].replace('has','')))
            item_state = gpt_call(get_prompt(object_state_prompt,"",character,sentence,slot['item needed'][0][1].replace('has','')), 12, 1, top_p=0.6, top_k=4)[0].split('\n')[0].replace('.','').strip()
            slot['item state'] = [(character,item_state)] if ('nothing' not in item_state and 'Nothing' not in item_state) else []
    if slot_flag['who']==True:
        #print(context_+get_prompt(parallel_precondition_second_character_prompt,"","",' '.join(sentence_.split(' ')).split('.')[0].split(' at ')[0]+' '+slot['location'][0]+'.'))
        for i in range(3):
            if slot_flag['location']==True and slot['location']!=[]:
                who_slot = gpt_call(context_+get_prompt(parallel_precondition_second_character_prompt,"","",' '.join(sentence.split(' ')).split('.')[0]+'.'), 13, 15, top_p=0.5, top_k=40)
                #who_slot = gpt_call(get_prompt(parallel_precondition_second_character_prompt,"","",' '.join(sentence_.split(' ')).split('.')[0].split('at')[0]+slot['location'][0]+'.'), 12, 15, temperature=0.8)
            else:
                who_slot = gpt_call(context_+get_prompt(parallel_precondition_second_character_prompt,"","",' '.join(sentence.split(' '))), 13, 15, top_p=0.5, top_k=40)
                #who_slot = gpt_call(get_prompt(parallel_precondition_second_character_prompt,"","",' '.join(sentence_.split(' '))), 12, 15, temperature=0.8)
            #print("debug who :",who_slot)
            slot['interaction with other person']=post_process(who_slot,prefix="",separator=';',must_not_contain = ')',
                                                    remove_first_word=False,mode="gen_actions",to_avoid = ' '.join(sentence.split(' ')),lookup={},simi_threshold=0.95,freq_threshold=15)
            if slot['interaction with other person']!=[]:
                print("debug interaction with other person :",slot['interaction with other person'])
                #slot['interaction with other person'][0]=slot['interaction with other person'][0]+' '+slot['location'][0]
                #slot['interaction with other person'] = [interaction.strip() for interaction in slot['interaction with other person'][0].split('.')]
                slot['interaction with other person'] = [(character,interaction) for interaction in slot['interaction with other person'] if len(interaction)>0]
                break
    if slot_flag['why']==True:
        reason = gpt_call(context_+get_prompt(parallel_reason_prompt,"","",' '.join(sentence.split(' ')).replace('.','')), 12, 15, top_p=0.4, top_k=20)
        slot['reason']=post_process(reason,remove_first_word=False,mode="gen_actions",to_avoid = sentence,action=sentence, prefix="",separator='.',freq_threshold=15,simi_threshold=0.94)
        if len(slot['reason'])>0:
            reasoning_graph.append([sentence_.strip().replace('.',''),slot['reason'][0]])
            slot['reason'] = [(character,slot['reason'][0])]
        #yield reason[0].strip(),"To "+end+", the person need to "+reason[0]+", which is solved by "+reason[0]+".\n",already_satisfied_preconditions
    return slot_flag, slot, sentence, character

def gpt_planner_single_step(end,lookup={},initial_state=[]):
    '''
    See the documentation at the beginning of this cell
    data struction:
        knowledge_graph: list[list[tuple(action, elaboration of action), tuple(precondtion, type of precondition), solution, character name]] 
        banned_preconditions_map: map{tuple(string,string)->list[tuple(string)]} [tuple(action,character):list[tuple(banned precondtions,character)]]
        knowledge_graph: list [tuple(action, elaboration of action), tuple(precondtion, type of precondition), solution, character name]
        open_conflict: list[list[action, elaboration of action, precondtion, type of precondition, character name]] 
        reasoning_graph: list[list[event, reason]]
        reusable_solutions: map{string->list[string]} [precondition: list of buffered candidate solutions]
    param end: the goal sentence
    param lookup: not used yet
    initial_state: list of string. Each item in initial_state is a clause that is satisfied for every characters by default. Example like "has money", "has car". It is used to end the recursion.
    '''
    global knowledge_graph, banned_preconditions_map, ends, open_conflict,reasoning_graph,reusable_solutions
    open_conflict=[]
    end_ = str(end)
    character = find_character_names(end)[0]

    #backtrack to produce context_ that help explain why current action is taken, also maintain banned_preconditions_map
    backtrack_result, num_item_state_in_the_plan = backtrack(end_.strip(),character)
    context_  = context_hint.replace('{$text1$}','\n'.join(['\n'.join([event[0]+' because '+event[1] for event in reasoning_graph]),backtrack_result]))
    last_precondition_type = None
    last_precondition = ''
    for entry in knowledge_graph:
        if end==entry[2]:
            last_precondition_type = entry[1][1]
            last_precondition = entry[1][0]

    #Fill slots (candidate preconditions)
    slot_flag, slots_filtered, end, character = get_slots_gpt(end,context_,last_precondition_type, num_item_state_in_the_plan=num_item_state_in_the_plan)

    #initialize already_satisfied_preconditions
    already_satisfied_preconditions = [(str(entry),1,character) for entry in initial_state]
    for entry in knowledge_graph:
        # we don't want the same character to perform the same action twice in a plan as this is a greedy planner
        already_satisfied_preconditions.extend([(entry[0][0],0,entry[3]),(entry[0][1],0,entry[3]),(entry[1][0],1,entry[3]),(entry[2],0,entry[3])])  
    already_satisfied_preconditions = list(set(already_satisfied_preconditions))
    print("iteration goal: %s"%end)

    #what preconditions are not allowed to appear for currect node. banned_preconditions_map is maintained by backtrack
    banned_preconditions = banned_preconditions_map[(end_.replace('.',''),character)]
    print('banned_preconditions: ',banned_preconditions)
    
    print("Elaborating:",end)
    print(slots_filtered)
    candidate_preconditions = []
    for key in slots_filtered:
        candidate_preconditions.extend([(precondition,key) for precondition in slots_filtered[key]])
    if len(candidate_preconditions)==0:
        "No action is selected"," finished"
    print("-----candidate_preconditions:")
    print(candidate_preconditions)

    #find open preconditions. Part I
    #detect conflict
    no_conflict_flag = True
    fact_about_object = ''
    if last_precondition_type=='item needed' and len(slots_filtered['location'])>0:
        fact_about_object,ownership_of_object = get_facts_and_check_ownership(end_,last_precondition.replace('has ',''),slots_filtered['location'][0][1])
    for (precondition,precondition_type) in copy.deepcopy(candidate_preconditions):
        no_conflict_flag = True
        if precondition_type=='negation':
            no_conflict_flag = False
        for banned_precondition in banned_preconditions:
            #print("debug banned precondition:",banned_precondition,precondition)
            if banned_precondition[-1]!=precondition[0]:
                continue
            similarity = cos_simi(precondition[1],banned_precondition[0])
            if last_precondition_type=='item needed' or last_precondition_type=='location':
                if banned_precondition[0].split(' ')[0]!=precondition[1].split(' ')[0]:
                    continue
                if similarity>0.75:
                    if last_precondition_type=='item needed' and cos_simi(precondition[1],last_precondition)>0.75:
                        #We recover some actions where the item belongs to the character because we want actions like "John get a book from the shelf at John's home" in our plan
                        #this helps to boost diversity.
                        if ownership_of_object==character:
                            print("**recovered a deletion based on facts")
                            print(end_,ownership_of_object,fact_about_object,precondition,last_precondition)
                            if (precondition,precondition_type) in candidate_preconditions:
                                candidate_preconditions.remove((precondition,precondition_type))
                            break
                    print("delete banned precondition 1: ",precondition,banned_precondition[0],similarity)
                    no_conflict_flag = False
                    break
            else:
                if precondition_type == 'interaction with other person' and similarity>0.9:
                    if (precondition,precondition_type) in candidate_preconditions:
                        candidate_preconditions.remove((precondition,precondition_type))
                elif similarity>0.9:
                    print("delete banned precondition 2: ",precondition,banned_precondition[0],similarity)
                    no_conflict_flag = False
                    break
        if no_conflict_flag==False and 'want' not in precondition[1] and 'need' not in precondition[1]:
            #if action is banned, we delect this action (entire node), gather its precondtions, find another way to satisfy the precondtion
            candidate_preconditions = []
            #delete previous action, collect entry from knowledge graph that starts with this action and collect entries whose solution is this action, add unsolved precondition caused by the deletion to the pending Q.
            index_to_remove = [idx for idx,entry in enumerate(knowledge_graph) if entry[0][0]==end_]
            temp = copy.deepcopy(knowledge_graph)
            knowledge_graph = [copy.deepcopy(temp[i]) for i in range(len(temp)) if i not in index_to_remove]
            reasoning_graph = [reason for reason in reasoning_graph if (end_.strip().replace('.','')!=reason[0] and end_.strip().replace('.','')!=reason[1])]
            index_to_pend_and_remove = [idx for idx,entry in enumerate(knowledge_graph) if entry[2]==end_]
            new_precondition_unsolved = [(knowledge_graph[i][0][0],knowledge_graph[i][0][1],knowledge_graph[i][1][0],knowledge_graph[i][1][1],
                                          knowledge_graph[i][3]) for i in index_to_pend_and_remove]
            open_conflict.extend(new_precondition_unsolved) 
            for i in range(len(new_precondition_unsolved)):
                print("remove similar solution from reusable_solutions",new_precondition_unsolved[i][2],end_[5:].strip().replace('.',''))
                to_remove_action = end_.strip().replace('.','')
                if 'has ' in new_precondition_unsolved[i][2] or 'at ' in new_precondition_unsolved[i][2]:
                    to_remove_action = ' '.join(to_remove_action.split(' ')[1:])
                reusable_solutions[(new_precondition_unsolved[i][2].strip(),new_precondition_unsolved[i][4])] = [solution for solution in reusable_solutions[(new_precondition_unsolved[i][2].strip(),new_precondition_unsolved[i][4])] if cos_simi(solution, to_remove_action)<0.75]
            temp = copy.deepcopy(knowledge_graph)
            if precondition_type!='negation':
                knowledge_graph = [copy.deepcopy(temp[i]) for i in range(len(temp)) if i not in index_to_pend_and_remove]
            else:
                print("add negation",precondition)
                knowledge_graph.append([(end_,end),(precondition,'negation'),'which is satisfied automatically.',character])

            print("add unsolved precondition: ",new_precondition_unsolved)
            break
         
    #print("-----already_satisfied_preconditions:")
    #print(already_satisfied_preconditions)
    
    for precondition,precondition_type in candidate_preconditions:
        no_match_flag = True
        to_merge_node = None
        for already_satisfied_precondition in already_satisfied_preconditions:
            if already_satisfied_precondition[-1]!=precondition[0]:
                continue
            similarity = cos_simi(precondition[1],already_satisfied_precondition[0])
            #merge duplicate precondition
            if 'has '==precondition[1][:4] or 'at '==precondition[1][:3]:
                if similarity>0.75:
                    print("merge duplicate precondition 1: ",precondition[1],already_satisfied_precondition[0],similarity)
                    no_match_flag = False
                    to_merge_node = copy.deepcopy(already_satisfied_precondition)
                    break
            else:
                if similarity>0.9:
                    print("merge duplicate precondition 2: ",precondition[1],already_satisfied_precondition[0],similarity)
                    no_match_flag = False
                    to_merge_node = copy.deepcopy(already_satisfied_precondition)
                    break
        if no_match_flag and 'want' not in precondition[1] and ('need' not in precondition[1] or precondition_type=='item state'):
            open_conflict.append((end_,end,precondition[1],precondition_type,precondition[0]))
        else:
            #fix conflict caused by merging
            #forward_fix_banned_preconditions(precondition[1],list(set(banned_preconditions+[(precondition[1],precondition[0])])))
            if to_merge_node:
                knowledge_graph.append([(end_,end),(to_merge_node[0],precondition_type),'which is satisfied automatically.',precondition[0]])
            else:
                knowledge_graph.append([(end_,end),(precondition[1],precondition_type),'which is satisfied automatically.',precondition[0]])
            yield "","","",True
    
    
    if fact_about_object!='' and len(candidate_preconditions)>0:
        knowledge_graph.append([(end_,end),(fact_about_object,'fact_about_object'),'which is satisfied automatically.',character])
    '''if len(candidate_preconditions)>0 and (last_precondition_type=='item needed' or last_precondition_type=='location'):
        double_negation = get_double_negation(end_)[0]
        knowledge_graph.append([(end_,end),(double_negation,'fact_about_object'),'which is satisfied automatically.',character])'''
    open_conflict = list(set(open_conflict))
    print("open conflict:",open_conflict)

    #knowledge_graph = [entry for entry in knowledge_graph if entry[1][1]!='negation' or cos_simi(entry[0][0],end_)<0.7]

    #find next actions (solutions) to take given open preconditions
    #actions for "item needed" are buffered and will be reused if the action is delected for causing a conflict
    while len(open_conflict)!=0:
        original_sentence, expanded_sentence,selected_precondition,selected_precondition_type, character_  = open_conflict[0]
        open_conflict.pop(0)
        print("selected_precondition for (%s):"%original_sentence)
        print(selected_precondition+' to '+' '.join(original_sentence.split(' ')[1:]))
        candidate_actions_filtered = []
        #print(reusable_solutions,(selected_precondition.strip(), character_))
        print("buffered solution for "+character_+" "+selected_precondition+":",reusable_solutions[(selected_precondition.strip(), character_)])
        if len(reusable_solutions[(selected_precondition.strip(), character_)])>0:
            if selected_precondition_type=='interaction with other person':
                continue
            candidate_actions_filtered = copy.deepcopy(reusable_solutions[(selected_precondition.strip(), character_)])
        else:
            satisfy_flag=False
            print("-----candidate_actions:")
            while satisfy_flag==False:
                if selected_precondition_type=='item needed' or selected_precondition_type=='location' or selected_precondition_type=='item state':
                    if selected_precondition_type=='item needed':
                        gpt_generated_actions = gpt_call(get_prompt(action_prompt_obtain_item,' '.join(expanded_sentence.split(' ')[0:]).strip(),character_,preprocess(selected_precondition+'?')),11, 20, temperature=0.9)
                    if selected_precondition_type=='location':
                        if any([key_word in selected_precondition for key_word in ['stand','station','stop']]) and any([key_word in selected_precondition for key_word in ['bus','taxi','subway']]):
                            gpt_generated_actions = [character_+' walked to get to '+selected_precondition.replace('at ','')+'.',character_+' drove to get to '+selected_precondition.replace('at ','')+'.']
                        else:
                            gpt_generated_actions = gpt_call(get_prompt(action_prompt_go_to_somewhere,' '.join(original_sentence.split(' ')[0:]).strip(),character_,preprocess(selected_precondition+'?')),11, 5, temperature=0.6)

                    if selected_precondition_type=='item state':
                        gpt_generated_actions = gpt_call(get_prompt(object_state_to_action_prompt,"",character_,original_sentence,selected_precondition),12, 5, temperature=0.75)
                    candidate_actions = post_process(gpt_generated_actions,';',
                                        mode="gen_actions",to_avoid = ' '.join(original_sentence.split(' ')[1:]),lookup=lookup,action='', simi_threshold=0.8,freq_threshold=2,exclude_go_to=not ' at ' in selected_precondition)
                
                    if 'at ' in selected_precondition:
                        candidate_actions = [action.split(' to ')[0]+' to get to '+' '.join(selected_precondition.split(' ')[1:]) for action in candidate_actions]
                else:
                    candidate_actions = [selected_precondition]
        
                if len(candidate_actions)==0:
                    continue
            
                candidate_actions_filtered = []
                print("debug candidate_actions:",candidate_actions)
                for candidate_action in candidate_actions:
                    if selected_precondition[0:3]=='has': 
                        candidate_actions_filtered.append(fix_missing_object(candidate_action,selected_precondition[4:]))
                    elif selected_precondition[0:2]=='at':
                        candidate_actions_filtered.append(fix_missing_object(candidate_action,selected_precondition[3:]))
                    else:
                        candidate_actions_filtered.append(candidate_action)
                if len(candidate_actions_filtered)==0:
                    continue
                else:
                    satisfy_flag = True
                    break
            reusable_solutions[(selected_precondition.strip(), character_)] = copy.deepcopy(candidate_actions_filtered)
        print("candidate_actions_filtered:",candidate_actions_filtered)
        selected_action = choose_best_action_by_perplexity(selected_precondition,candidate_actions_filtered).strip()
        print("selected_action for (%s):"%selected_precondition)
        if selected_precondition_type=='interaction with other person':
            character_ = find_character_names(selected_action)[0]
        elif character_!=selected_action.split(' ')[0]:
            selected_action = character_+' '+selected_action
        selected_action = selected_action.replace('.','')
        print(selected_action)
        knowledge_graph.append([(original_sentence,expanded_sentence),(selected_precondition,selected_precondition_type),selected_action+'.',character_])
        yield selected_action, character_,selected_precondition_type,no_conflict_flag

def iterative_gpt_planner(end,iterations=10,initial_state=['has car']):
    '''
    See the documentation at the beginning of this cell
    param end: the goal sentence
    param iterations: how many iterations until we cut the recursion
    initial_state: list of string. Each item in initial_state is a clause that is satisfied for every characters by default. Example like "has money", "has car". It is used to end the recursion.
    '''
    global text, banned_preconditions_map, knowledge_graph, reasoning_graph,ends,reusable_solutions
    reusable_solutions = defaultdict(list)
    text = ""
    end_ = str(end)
    ends = [end]
    banned_preconditions_map = defaultdict(list)
    banned_preconditions_map[end]=[]
    knowledge_graph = []
    reasoning_graph = []
    terminate = False
    for iter in range(iterations):
        print("knowledge graph:")
        print(knowledge_graph)
        new_ends = []
        time_out = 0
        while len(ends)!=0:
            print('****edges: ',ends)
            end = ends.pop(0)
            planner = gpt_planner_single_step(end,initial_state=initial_state)
            out = list(planner)
            if len(out)==0:
                continue
            for o in out:
                if o[-1]==False:
                    if time_out<=8:
                        if o[2]=='interaction with other person':
                            ends.insert(0,(o[0]+'.').strip())
                        else:
                            if  o[1]+' ' not in o[0]:
                                ends.insert(0,(o[1]+" "+o[0]+'.').strip())
                            else:
                                ends.insert(0,(o[0]+'.').strip())
                        time_out+=1
                    else:
                        time_out=0
                elif len(o[0])>0:
                    new_ends.append(o)
        ends = []
        print("--------Iteration %i--------"%iter)
        for action,character,action_type,no_conflict_flag in new_ends:
            if len(action)!=0:
                if action_type=='interaction with other person':
                    ends.append((action+'.').strip())
                else:
                    if  character+' ' not in action:
                        ends.append((character+" "+action+'.').strip())
                    else:
                        ends.append((action+'.').strip())
        ends = list(set(ends))
        print(ends)
        if len(ends)==0:
            break
        if terminate:
            break
    text = forward_traverse(end_)
    print("banned precondition map:")
    print(banned_preconditions_map)
    print("full text:")
    print(text)
    print("\n\nfinal text:")
    [print(s) for s in text.split('\n') if 'automatic' not in s]
    return text.split('\n'),copy.deepcopy(knowledge_graph)
#iterative_gpt_planner("John gave Sally a book.",iterations=5)
```

```python
def get_double_negation(sentence, num_generations = 1, verbose = False):
    neg_output = negate_input(sentence, num_generations = num_generations)

    if verbose:
      print("input negation: ", neg_output)

    cfp_gens = counter_fact_precond(neg_output[0][:-1], num_generations = num_generations)

    uniq_cfp_gens, counts = np_unique_sent(cfp_gens, return_counts = True)
    uniq_cfp_gens, counts = list_sort(uniq_cfp_gens, counts)
    uniq_cfp_gens.reverse()
    counts.reverse()

    if verbose:
      print("cfpreconditions: ", uniq_cfp_gens)
      print("counts: ", counts)

    negations = []

    for gen in uniq_cfp_gens:
      all_negations, neg_counts = np_unique_sent(negate_input(gen), return_counts = True)
      all_negations, neg_counts = list_sort(all_negations, neg_counts)

      if verbose:
        print("all candidate negations for ", gen)
        print(all_negations)
        print(neg_counts)

      negations.append(all_negations[np.argmax(neg_counts)])
    
    return negations

def negate_input(sentence, num_generations = 5):
  '''
  return the negation of the input sentence
  param sentence: string. the action or statement
  param generations: int. the number of times to query GPT
  '''

  negation_prompt_step1 = '''Hanna got in the car. The negation of this is Hanna didn't get in the car.
  Steve went to school. The negation of this is Steve didn't go to school.
  Bob gave Jenny a key. The negation of this is Bob didn't give Jenny a key.
  Timmy didn't have a gun. The negation of this is Timmy had a gun.
  {$text1$} The negation of this is'''

  all_gens = []
  prompt = get_prompt(negation_prompt_step1, sentence)

  while len(all_gens) < num_generations:
    gpt_output = gpt_call(prompt, 10, 15, top_p=0.6, top_k=10, return_probability=False)
    output           = gpt_output[0].split("\n")[0]
    # print("output: ", output)
    for atom in split_sentence_array(output):
      if len(atom) > 1 and ("_" not in atom):
        all_gens.append(atom)
  
  return all_gens


def counter_fact_precond(sentence, num_generations = 5):
  '''
  Given a statement, expand on the cause or potential reason behind the statement.
  param sentence: string. the action or statement.
  param generations: int. the number of times to query GPT
  '''

  negation_prompt_step2 = '''Hanna didn't give Bob an apple because Bob was allergic to apples.
  Steve didn't go to the park. This was because the park was closed.
  Jen didn't go into her house. This was because Jen didn't have her keys.
  Ben didn't go to school. This was because Ben was sick.
  Ethan didn't drive to work. This was because Ethan's car was out of gas.
  Jake wasn't hungry. This was because Jake ate a burger.
  {$text1$}. This was because'''

  all_gens = []
  prompt = get_prompt(negation_prompt_step2, sentence)

  max_attempts = 10
  attempts     = 0

  while len(all_gens) < num_generations and attempts < max_attempts:
    gpt_output = gpt_call(prompt, 10, 15, top_p=0.6, top_k=10, return_probability=False)
    output           = gpt_output[0].split("\n")[0]
    # print("output: ", output)
    for atom in split_sentence_array(output):
      if len(atom) > 1 and ("_" not in atom):
        all_gens.append(atom)

    attempts +=1
    
  return all_gens

print(get_double_negation("Tim ran a marathon.", num_generations = 3, verbose = True))
```

```python
def gpt_call(prompt,max_token,num_of_sentences,temperature=0.9, top_p=None, top_k=None, return_probability=False):
    '''
    GPT-J API:
        returning multiple generated sentence if return_probability is False
        returning list of (samples, probability, tokens, logits) if return_probability is True. 
            samples is the generated sentence in text by default sampling method. (in this colab demo, batch size is 1, so only one sentence is returned)
            probability is a list of probability for each word in samples
            tokens is a list of tokens for each word in samples
            logits: list of numpy vector. posterior probability at each step given the previous tokens. It is probability not logits!!!!
    param context: prompt
    param num_of_sentences: number of sentence to generate
    param top_k (int, optional, defaults to model.config.top_k or 50 if the config does not set any value) — The number of highest probability vocabulary tokens to keep for top-k-filtering.
    param top_p (float, optional, defaults to model.config.top_p or 1.0 if the config does not set any value) — If set to float < 1, only the most probable tokens with probabilities that add up to top_p or higher are kept for generation.
    param temperature:  (float, optional, defaults to model.config.temperature or 1.0 if the config does not set any value) — The value used to module the next token probabilities.
    param max_token: num of tokens to generate
    param return_probability: if set to True, return list of (samples, probability, tokens, logits)
    '''
    ret = []
    ret_prob = []
    ret_tokens = []
    ret_logits = []
    if top_p!=None and top_k!=None:
        for i in range(num_of_sentences):
            result = infer(prompt, top_p=top_p, top_k=top_k,  temp=temperature, gen_len=max_token,return_probability=return_probability)
            ret.append(result[0])
            if return_probability:
                ret_prob.append(result[1])
                ret_tokens.append(result[2])
                ret_logits.append(result[3])
    else:
        for i in range(num_of_sentences):
            result = infer(prompt, top_p=top_p, top_k=top_k, temp=temperature, gen_len=max_token, return_probability=return_probability)
            ret.append(result[0])
            if return_probability:
                ret_prob.append(result[1])
    if return_probability:
        return ret, ret_prob, ret_tokens,ret_logits
    return ret
```

## Neural Planner



### Workflow:

* nodes are in the form of **(character, action, precondition, solution)**
* Given: an action as a sentence s: 

* Create an empty node N0 (XX, s, XX, XX). add it to the empty graph G: 

* Put N0 in Q, a priority queue 

* While Q is not empty: 
>* N = pop Q 

>* Filling Slots & Use GPT to make precondition questions & Make a list of candidate preconditions base on N
>>* fill the character slot (c) by prompting GPT-J
>>* Perform syntax analysis on s.
>>* (optional) Use few-shot learning to teach GPT-J to expand the syntax and include keywords if possible. 
>>* if a keyword shows up in the expanded syntax, it means we need to ask certain type of questions.
>>>* For instance, "John was arrested"->"somebody was arrested"->"somebody was arrested `because` somebody did something"-> 'because' detected -> ask "why John was arrested?"
>>* Answer selected questions with GPT-J to fill in preconditions
>>* We have 4 types of questions and thus 4 types of preconditions. 
>>> `item needed`, `interaction with person`, `location`, `reason`
>* For each character preconditions pair (c,q). Determine if it is an open precondition
>>* If (c,q) is satified by default, ignore it.
>>* If (c,q) has already appeared somewhere in the path to N, discard node N, gather all unsolved preconditions caused by the deletion of N and add those unsolved preconditions to open conflict list.
>>* If (c,q) has already appeared somewhere but not in the path to N. Merge two branchs to reuse actions. Check forward to detect and remove all potential conflicts. 
>>* otherwise, append (c,q) to open-conflict slot. 

>* For each open precondition (c,p)
>>* generate a solution a with GPT-J
>>>* we only generate solutions for preconditions of type `item needed`, and `location`. We use the precondition directly as solutions for those of type `interaction with person`, and `reason`.
>>>* We buffer the generated solution so that we can reuse them if a solution is rejected due to confliction
>>>* We only generate solutions if we run out of buffered solutions.
>>* Duplicate node N. Add a new node N'=(character, s, p, a) to G
>>* Create a new node (XX, a, XX, XX). Push it to Q

```python
#from sys import last_traceback


import spacy
import copy
NLP = spacy.load("en_core_web_sm")


lemmatizer = nltk.stem.WordNetLemmatizer()


parallel_precondition_item_prompt = '''Find out what item is required for the given circumstance, if nothing is needed, answer nothing.

What item must {$text4$} possess in order to clean table?
Answer: {$text4$} must have a table cloth,

What item must {$text4$} possess to rode his bike with friends?
Answer: {$text4$} must have a bike,

What item must {$text4$} possess to buy book from the book store?
Answer: {$text4$} must have money,

What item must {$text4$} possess to run away from a bear?
Answer: {$text4$} musto have nothing,

What item must {$text4$} possess to gave his parent a birthday gift?
Answer: {$text4$} must have a birthday gift,

What item must {$text4$} possess to shoot someone?
Answer: {$text4$} must have a gun,

What item must {$text4$} possess to watch a movie at the cinema?
Answer: {$text4$} must have a movie ticket,

What item must {$text4$} possess to was sitting on a beach?
Answer: {$text4$} must have sunglasses,

What item must {$text4$} possess to walk to the bus station?
Answer: {$text4$} must have nothing,

What item must {$text4$} possess to take a taxi?
Answer: {$text4$} must have money,

What item must {$text4$} possess to bought a bundle of flower?
Answer: {$text4$} must have money,

What item must {$text4$} possess to got ticket from ticket office?
Answer: {$text4$} must have money,

What item must {$text4$} possess to {$text3$}?
Answer: {$text4$} must have{$text2$}'''


how_to_get_to_goal_prompt = '''Given the sentence, answer how the character achieves the goal.

Sentence: Tim became a member of the school basketball team.
How does Tim became a member of the school basketball team?
Tim practices basketball everyday.

Sentence: Millian shot Mike.
How does Millian shot Mike?
Millian shot Mike.

Sentence: Jonny was prompted.
How does Jonny was prompted?
Jonny works hard and has excellent performance.

Sentence: Fred became famous.
How does Fred became famous?
Fred wrote a book and it sold well.

Sentence: Willson died of illness.
How does died of illness?
Willson died of illness.

Sentence: Tony got an A for a class.
How does Tony got an A for a class?
Tony cheated in the exam.

Sentence: Tom works hard at night.
How does Tom works hard at night?
Tom works hard at night.

Sentence: {$text3$}.
How does {$text3$}?
'''


parallel_precondition_action_prompt = '''How to buy something?
Answer: by ordering online;

How to get a good grade in an exam?
Answer: by studying hard

How to go to home?
Answer: by taking a bus;

How to {$text3$}?
Answer: by{$text2$}'''



parallel_precondition_second_character_prompt = '''What happens between the protagonist and another character prior to the event?
Event: Tom bought a second-hand car.
prior to Tom bought a second-hand car, Tom negotiated with a used car dealer;

Event: Tim chats with Smith at a bar.
prior to Tim chatted with Smith at a bar, Tim met Smith at the bar;

Event: Danie had a fight with Tony.
prior to Danie had a fight with Tony, Danie had an argument with Tony;

Event: Fred hugs Tom on the street.
prior to Fred hugged Tom, Fred met Tom on the street;

Event: {$text3$}.
prior to {$text3$},'''


parallel_precondition_location_prompt = '''Josh buys rice from the grocery at grocery.
Where is Josh? grocery;

Sam meets Fred at the museum.
Where is Sam? museum;

Tom watched a movie.
Where is Tom? cinema;

Jack takes a suit from the wardrobe.
Where is Jack? home;

Fred takes an airplane.
Where is Fred? airport;

Julia sees a bear.
Where is Julia? forest;

Thomas takes a kichen knife from the kitchen.
Where is Thomas? kitchen;

Jim bought a bottle of wine.
Where is Jim? supermarket;

Amy buys a fishing pole from a fishing store.
Where is Amy? fishing store;

Bob bought a new pen.
Where is Bob? stationer;

Tony bought a bottle of water for his wife.
Where is Tony? shop;

{$text2$} {$text3$}.{$text4$}
Where is {$text2$}?'''

parallel_effect_prompt = '''How does a person have some food to eat food?
Answer: go to a restaurant; order takeout online；

How does a person study to get good grades in an exam?
Answer: revise to prepare for the exam;

How does a person by taking the bus to go to the museum?
Answer: take the bus;

How does a person at home to party with friends?
Answer: drive home; walk home;

how does a person {$text3$}?
Answer:{$text2$}'''


action_prompt_obtain_item = '''Context: {$text2$} eats a hamburger.
How did {$text2$} get a hamburger?
Answer: {$text2$} bought a hamburger at a restaurant.

Context: {$text2$} opened the envelope.
How did {$text2$} get a knife?
Answer: {$text2$} gets a knife from the drawer.

Context: {$text2$} takes a shower.
How did {$text2$} get a towel?
Answer: {$text2$} gets a towel from towel rack.

Context: {$text2$} watch a TV show.
How did {$text2$} get TV?
Answer: {$text2$} buys a TV from the supermarket.

Context: {$text2$} went to a concert.
How did {$text2$} get a ticket?
Answer: {$text2$} buys a ticket from the ticket office.

Context: {$text1$}
How did {$text2$} {$text3$}
Answer: {$text2$}'''

action_prompt_go_to_somewhere = '''Context: {$text2$} eats a hamburger at restaurant.
How did {$text2$} get to restaurant?
Answer: {$text2$} drove to get to the restaurant.

Context: {$text2$} had a stroke.
How did {$text2$} get to hospital?
Answer: {$text2$} was taken to the hospital by an ambulance.

Context: {$text2$} had a party at home.
How did {$text2$} get home?
Answer: {$text2$} walked home.

Context: {$text2$} greets his teacher at the classroom.
How did {$text2$} get to school?
Answer: {$text2$} took a bus to get to the school.

Context: {$text1$}
How did {$text2$} {$text3$}
Answer: {$text2$}'''

parallel_reason_prompt = '''Explain what the protagonist has done that causes the event to happen.

Example 1: John went to the hospital because Mike wanted to see a doctor at the hospital.

Example 2: Fred got a zero on the exam because Fred cheated during the exam at school.

Example 3: Tom reads a book because Tom wants to read a book.

Example 4: Tina was fired by her boss because Tina failed to finish her work at the company.

Example 5: Thomas takes the bus to get to the library because Thomas wants to go to the library.

Example 6: Julia broke up with her boyfriend because Julia's boyfriend didn't reply to her message.

Example 7: Tim had an amputation because he was hit by a car badly.

Example 8: {$text3$} because{$text2$}'''

enrich_context_prompt='''Elaborate the given event to include a location where the event takes place.

Original sentence: Tony watched a movie.
Sentence with a location where the event takes place: Tony watched a movie at the cinema.

Original sentence: Tina bought a pancake at the supermarket.
Sentence with a location where the event takes place: Tina bought a pancake from a supermarket.

Original sentence: Mike went hiking.
Sentence with a location where the event takes place: Mike went hiking with his friends at a national park.

Original sentence: Jam was seen drinking at a bar by the bartender.
Sentence with a location where the event takes place: Jam was seen drinking at a bar by the bartender.

Original sentence: {$text3$}
Sentence with a location where the event takes place: {$text2$}'''


extract_character_name_prompt='''Name the principal character in the sentence. Here are some examples:
Example 1:
Sentence: "Hi, Tim", Mark said to Tim.
Characters: Mark,

Example 2: 
Sentence: Sherry went to bed.
Characters: Sherry,

Example 3:
Sentence: Jack met with Sam.
Characters: Jack

Example 4:
Sentence: the lawyer had an arguement with his wife.
Characters: the lawyer

Example 5:
Sentence: {$text3$}
Characters:'''

object_state_to_action_prompt = '''Given the sentence, what we tell about the state of the item in the sentence?

Context: Tim picked up a bottle of water from the table.
The bottle needs to be full.
Tim takes water from the tap at the restroom.

Sentence: Millian shot Mike on the street.
The gun needs to be loaded.
Millian loaded the gun on the street.

Sentence: Jonny took a bus to get home.
The bus ticket needs to be validated.
Jonny let the ticket collector validate the ticket.

Sentence: Fred made steak in the kitchen.
The steak needs to be marinated.
Fred marinated the steak in the kitchen.

Sentence: Willson drove his car to Sally's home.
The car needs to be repaired.
Willson repairs his car at a garage.

Sentence: Tony wrote with a typewriter.
The typewriter needs to be turned on.
Tony turned on the typewriter.

Sentence: {$text3$}
{$text4$}
{$text2$}'''

object_state_prompt = '''Given the sentence, what we tell about the state of the item in the sentence?

Sentence: Tim picked up a bottle of water from the table.
What can we tell about the bottle of water other than how Tim obtain it?
nothing.

Sentence: Millian shot Mike.
What can we tell about the gun other than how Millian obtain it?
The gun needs to be loaded.

Sentence: Jonny took a bus to get home.
What can we tell about the bus ticket other than how Jonny obtain it?
nothing.

Sentence: Fred made steak.
What can we tell about the steak other than how Fred obtain it?
The steak needs to be marinated.

Sentence: Willson drove his car to Sally's home.
What can we tell about the car other than how Willson obtain it?
nothing.

Sentence: Tony wrote with a typewriter.
What can we tell about the typewriter other than how Tony obtain it?
The typewriter needs to be turned on.

Sentence: Tom paid his rent.
What can we tell about the money other than how Tom obtain it?
nothing.

Sentence: {$text3$}
What can we tell about the {$text4$} other than how {$text2$} obtain it?
'''

from collections import defaultdict
text = ""
knowledge_graph = []
reasoning_graph = []
banned_preconditions_map = defaultdict(list)
ends = []
open_conflict = []
reusable_solutions = defaultdict(list)
context_hint = '''Here is some context for the event that's going to happen.
{$text1$}

'''

3
def get_prompt(prompt,text1="",text2="", text3="", text4="", text5=""):
    '''
    generate prompt with templates
    '''
    if "{$text1$}" in prompt:
        prompt = prompt.replace("{$text1$}",text1)
    if "{$text2$}" in prompt:
        prompt = prompt.replace("{$text2$}",text2)
    if "{$text3$}" in prompt:
        prompt = prompt.replace("{$text3$}",text3)
    if "{$text4$}" in prompt:
        prompt = prompt.replace("{$text4$}",text4)
    if "{$text5$}" in prompt:
        prompt = prompt.replace("{$text5$}",text5)
    return prompt


def preprocess(candidate_preconditions):
  '''
  preprocess before generating actions for preconditions
  '''
  if isinstance(candidate_preconditions,str):
    if "has done" in candidate_preconditions:
      candidate_preconditions = candidate_preconditions.replace("has done ","")
    if "has " in candidate_preconditions:
      candidate_preconditions = candidate_preconditions.replace("has ","get ")
    if "at " in candidate_preconditions:
      candidate_preconditions = candidate_preconditions.replace("at ","get to ")
    return candidate_preconditions
  else:
    ret = []
    for precondition in candidate_preconditions:
      if "has done" in precondition:
        precondition = precondition.replace("has done ","")
      if "has " in precondition:
        precondition = precondition.replace("has ","get ")
      if "at " in precondition:
        precondition = precondition.replace("at ","get to ")
      ret.append(precondition)
    return ret

def lemmarization(word):
    return lemmatizer.lemmatize(word)

def match_lemma(precondition1,precondition2,threshold=0.6):
    word_lemma1 = [lemmarization(word) for word in precondition1.split(' ')]
    word_lemma2 = [lemmarization(word) for word in precondition2.split(' ')]
    
    u=len(set(word_lemma1).union(set(word_lemma2)))
    i = len(word_lemma1)+len(word_lemma2)-u
    if i/u>threshold:
        return True
    else:
        return False

def get_verb(sentence):
  doc = NLP(sentence)
  for token in doc:
    if token.pos_ == 'VERB':
      return token.text, token.lemma_
  return "",""

def choose_best_action_by_perplexity(effect,actions):
    true_false_prompt = '''Determine if the statement is true or false

    statement: at home is the result of John drove to get to home. 
    answer: true

    statement: {$effect$} is the result of {$action$} 
    answer:'''
    filtered_actions = []
    for action in actions:
        res = gpt_call(true_false_prompt.replace("{$action$}",action).replace("{$effect$}",effect),1,1,return_probability=True)
        text = res[2][0][0][1]
        prob = res[1][0][0]
        if text==' true':
            filtered_actions.append((action,prob))
    scores = sorted(filtered_actions,key=lambda x:x[1])
    print(scores)
    return scores[0][0]

def choose_best_action_by_perplexity(effect,actions,parent_action,character):
    true_false_prompt_precondition_satisfied = '''Determine if the statement is true or false

statement: John drove to get to home. John at school
answer: false

statement: John bought milk from the store. John has milk
answer: true

statement: John took bus to get to company. John at company
answer: true

statement: John read a book. John has book 
answer: true

statement: {$action$} {$effect$}
answer:'''
    true_false_prompt_is_coherent = '''Determine if the two sentences are logical coherent.

sentences: John walked home. John have a party at home. 
Are those sentences logical coherent? true

sentences: John got kicked out of the college. John got a college degree. 
Are those sentences logical coherent? false

sentences: {$parent_action$}. {$child_action$}. 
Are those sentences logical coherent?'''
    '''
    filtered_actions = []
    print("filtering actions: ",actions)
    for action in actions:
        #print(true_false_prompt_precondition_satisfied.replace("{$action$}",action).replace("{$effect$}",character+' '+effect))
        res = gpt_call(true_false_prompt_precondition_satisfied.replace("{$action$}",action).replace("{$effect$}",character+' '+effect),1,1,return_probability=True,top_k=1,top_p=0.7)
        #print(res)
        text = res[0][0][0]
        prob = res[1][0][0]
        if text==' true':
            filtered_actions.append((action,prob))
    if filtered_actions==[]:
        return "",[]
    scores = sorted(filtered_actions,key=lambda x:x[1],reverse=True)
    print(scores)
    '''
    '''
    filtered_actions = []
    actions = [t[0] for t in scores]
    for action in actions:
        res = gpt_call(true_false_prompt_is_coherent.replace("{$parent_action$}",parent_action).replace("{$child_action$}",action),1,1,return_probability=True,top_k=1)
        text = res[0][0][0]
        prob = res[1][0][0]
        if text==' true':
            filtered_actions.append((action,prob))
    if filtered_actions==[]:
        return "",[]
    scores = sorted(filtered_actions,key=lambda x:x[1],reverse=True)
    print(scores)
    '''
    scores = [[action,score(action)] for action in actions]
    scores = sorted(scores,key=lambda x:x[1],reverse=True)
    return scores[0][0], [t[0] for t in scores]


def choose_best_precondition_by_perplexity(action,preconditions):
    '''
    return the best precondition for certain action according to perplexity
    param action: action
    param preconditions: candidate preconditions for the action
    '''
    true_false_prompt = '''Determine if the statement is true or false
    statement: One need to {$precondition$} to {$action$}
    answer: true'''
    scores = [score(true_false_prompt.replace("{$precondition$}",precondition).replace("{$action$}",action)) for precondition in preconditions]
    print(scores)
    return preconditions[scores.index(min(scores))]


def choose_precondition_by_frequency(preconditions, freq_threshold=2,cutoff_frequency=0,action="",match_lemma_threshold=0.66, use_majority_vote = True, majority_vote_threshold=0.5):
    '''
    choose most common preconditions from candidates, a candidate precondition is considered as valid once its frequency pass this threshold
    When none of candidate preconditions pass the threshold, the most frequent one will be returned if its frequency exeed cutoff_frequency.
    when action is given, return the best suit precondition given the action.
    param preconditions: candidate preconditions
    param freq_threshold: a candidate precondition is considered as valid once its frequency pass this threshold
    param cutoff_frequency: When none of candidate preconditions pass the threshold, the most frequent one will be returned if its frequency exeed cutoff_frequency.
    param action: when action is given, return the best suit precondition given the action.
    param match_lemma_threshold: between 0 and 1, a higher value indicate less tolerance in fuzzy matching
    param use_majority_vote: if True, return the majority vote result (More than 50% of votes)
    '''
    pool = {}
    for precondition in preconditions:
        pool[precondition] = 0
    for precondition in preconditions:
        for key in pool.keys():
            if match_lemma(precondition,key,threshold=match_lemma_threshold):
                pool[key]+=1
    most_common_freq = 0
    most_common_precondition = None
    ret = []
    pool = dict(sorted(pool.items(),key=lambda x:x[1],reverse=True))
    print(pool)
    for key in pool.keys():  
        if use_majority_vote and pool[key]>=len(preconditions)*majority_vote_threshold:
            return [key]  
        if pool[key]>=most_common_freq:
            most_common_precondition = key
            most_common_freq = pool[key]
        if pool[key]>=freq_threshold:
            ret.append(key)
    if most_common_precondition not in ret and most_common_freq>cutoff_frequency:
        ret.append(most_common_precondition)
    if action!="" and ret!=[]:
        ret=[choose_best_precondition_by_perplexity(action, ret)]
    return ret

def post_process(generated_sentences,separator=';',prefix='', mode="gen_preconditions", must_have='',must_not_contain = '', to_avoid = "",action="",remove_first_word=False,exclude_go_to=False,
                 lookup={},simi_threshold=0.8,freq_threshold=20,cutoff_frequency=0,match_lemma_threshold=0.66, use_majority_vote = True, majority_vote_threshold=0.5):
    '''
    all-in-one postprocess function for postprocessing outputs of GPT-J.
    param generated_sentences: original GPT-J output
    param seperator: a string for the delimiter
    param prefix: added as the prefix to the postprocess results
    param mode: string, if "gen_actions", filter all candidates that are similar to to_avoid
    param must_have: string, filter out candidates that doesn't contain it
    param to_avoid: string, must be non-empty if mode is "gen_actions"
    param action: for choose_best_precondition_by_perplexity
    remove_first_word: whether or not to remove the first word, usually the name of the protagonist
    param lookup: not used
    '''
    candidates = []
    if must_have!='':
        generated_sentences = [s for s in generated_sentences if must_have in s]
    #print(generated_sentences)
    for result in generated_sentences:
        result = result.split('\n')[0]
        if separator in result:
            result = [(s.strip()).strip() for s in result.split(separator)[:-1]]
        else:
            result = [(s.strip()).strip() for s in result.split(separator)]
        if mode=="gen_actions":
          for action_ in result:
            similarity_score = cos_simi(to_avoid,action_.split('.')[0])
            if similarity_score<simi_threshold and len(action_.replace('.','').strip().split(' '))>1:
                if ' is ' not in action_ and ' was ' not in action_:
                    candidates.append(action_)
        else:
          candidates.extend(result)
    candidates = [candidate.strip() for candidate in candidates]
    candidates = [candidate for candidate in candidates if len(candidate)>0]   #get rid of empty string
    
    if must_not_contain!='':
        candidates = [candidate for candidate in candidates if must_not_contain not in candidate]
    if len(candidates)==0:
        return candidates
    print("candidates",candidates)
    for i in range(len(candidates)):
        if candidates[i].split(' ')[0] in ['a','an','the'] or remove_first_word:
            candidates[i] = ' '.join(candidates[i].split(' ')[1:])
    if exclude_go_to:
        candidates = [candidate for candidate in candidates if 'go to' not in candidate and 'goes to' not in candidate]
    if mode=="gen_item":
        temp = copy.deepcopy(candidates)
        for i,item in enumerate(temp):
            for item2 in temp:
                if len(item)<len(item2) and item in item2:
                    candidates.append(item2)
    candidates = choose_precondition_by_frequency(candidates, freq_threshold=freq_threshold, cutoff_frequency =cutoff_frequency, 
                                                  action=action, match_lemma_threshold=match_lemma_threshold, use_majority_vote=use_majority_vote, majority_vote_threshold=majority_vote_threshold)
    if mode=="gen_item":
        candidates = [candidate for candidate in candidates if candidate not in ['nothing','something']]
    return [prefix+' '+candidate for candidate in candidates]

def fix_missing_object(sentence,item):
    '''
    replace any pronom with specific object
    '''
    tokens = nltk.word_tokenize(sentence)
    tagged = nltk.pos_tag(tokens)
    ret = ''
    for word in tagged:
        new_word = word[0]
        if word[1]=='PRP':
            new_word = item
        ret+=new_word+' '
    return ret.strip()

def backtrack(target,character):
    '''
    backtrack, return the logic chain (context) that explains why a target event happen
    param target: target event
    already_contains_item_state: the number of item state contraints in all sub causal chains, used to terminate recursion.
    '''
    global banned_preconditions_map
    #print(target,character,'.')
    ids = []
    new_targets = []
    logic_chain = ""  #"To John was arrested. (John was arrested at his house. because  John's wife saw him taking drugs) , the person need to has police, which is solved by was arrested at his house."
    visited = []
    already_contains_item_state = 0
    target_ = target.replace('.','')
    banned_preconditions_map[(target_,character)] = []
    if (target_,character) not in banned_preconditions_map[(target_,character)]:
        banned_preconditions_map[(target_,character)].append((target_,character))
    preconditions = []
    for entry in knowledge_graph:
        if entry[2]==target and entry[3]==character:
            preconditions.append(entry[1])
    preconditions = list(set(preconditions))
    explored = [target_,target]
    for entry in knowledge_graph:
        if entry[1] in preconditions and entry[0][0] not in explored and entry[3]==character:
            explored.append(entry[0][0])
            logic_chain_next, pending_number_item_state = backtrack(entry[0][0],find_character_names(entry[0][0])[0])
            already_contains_item_state += pending_number_item_state
            logic_chain+=logic_chain_next
            banned_preconditions_map[(target_,character)].extend(banned_preconditions_map[(entry[0][0].replace('.',''),entry[3])])
            banned_preconditions_map[(target_,character)].append((entry[1][0],entry[3]))
            banned_preconditions_map[(target_,character)] = list(set(banned_preconditions_map[(target_,character)]))
            if 'which is satisfied automatically.'!=entry[2] and entry[2]==target:
                logic_chain+='\nTo '+entry[0][0]+' ('+entry[0][1]+'), '+entry[3]+' need to '+entry[1][0]+', which is solved by '+entry[2]+'.'
                if entry[1][1]=='item state':
                    already_contains_item_state += 1

    return logic_chain, already_contains_item_state

    
def forward_traverse(end):
    '''
    Used for coverting a DAG to a plan in text
    param end: end node of the plan (end[0] is the input to the pipeline)
    '''
    edges = [end]
    ids = []
    while len(edges)>0:
        #print(edges)
        edge = edges.pop(0)
        for id,node in enumerate(knowledge_graph):
            if id not in ids and edge==node[0][0]:
                edges.append(node[3]+' '+node[2] if node[3]!=node[2].split(' ')[0].strip() else node[2])
                ids.append(id)
    formated_output = '\n'.join(['To '+knowledge_graph[id][0][0]+' ('+knowledge_graph[id][0][1]+'), '+knowledge_graph[id][3]+' need to '+knowledge_graph[id][1][0]+', which is solved by '+knowledge_graph[id][2] for id in ids])
    return formated_output  

def check_contains_upper(s):
    cnt = 0
    for c in s:
        if not c.isalpha():
            continue
        if cnt==0:
            if (not c.isalpha()) or c != c.upper():
                return False
        else:
            if c.isalpha() and c==c.upper():
                return False
        cnt+=1
    return True 

def find_character_names(sentence):
    '''
    return the principal character's name
    '''
    extract_principle_character_name_prompt='''Name the principal character in the sentence. Here are some examples:
Example 1:
Sentence: "Hi, Tim", Mark said to Tim.
Characters: Mark,

Example 2: 
Sentence: Sherry went to bed.
Characters: Sherry,

Example 3:
Sentence: Jack met with Sam.
Characters: Jack

Example 4:
Sentence: the lawyer had an arguement with his wife.
Characters: the lawyer

Example 5:
Sentence: {$text3$}
Characters:'''
    extract_all_characters_name_prompt='''Name all characters in the sentence. Here are some examples:
Example 1:
Sentence: "Hi, Tim", Mark said to Tim.
Characters: Mark, Tim,

Example 2: 
Sentence: Sherry went to bed.
Characters: Sherry,

Example 3:
Sentence: Jack met with Sam.
Characters: Jack, Sam,

Example 4:
Sentence: the lawyer had an arguement with his wife.
Characters: the lawyer, the lawyer's wife,

Example 5:
Sentence: Mike works hard at Apple Inc.
Characters: Mike,

Example 6:
Sentence: {$text3$}
Characters:'''
    prompt = get_prompt(extract_principle_character_name_prompt,"",'',sentence)
    output = gpt_call(prompt, 6, 1, top_p=0.5, top_k=1)
    #print(output)
    major_characters = post_process(output,prefix="",mode="gen_item",freq_threshold=0, cutoff_frequency=0,  match_lemma_threshold=0.9,separator=',')

    prompt = get_prompt(extract_all_characters_name_prompt,"",'',sentence)
    output = gpt_call(prompt, 6, 1, top_p=0.5, top_k=1)
    #print(output)
    all_characters = post_process(output,prefix="",mode="gen_item",freq_threshold=0, cutoff_frequency=0,  match_lemma_threshold=0.9,separator=',',use_majority_vote=False)
    all_characters = [ch.strip() for ch in all_characters if check_contains_upper(ch) and "'s" not in ch]
    if len(major_characters)==0:
        return "PersonX"
    else:
        return major_characters[0].strip(),all_characters
            

def forward_fix_banned_preconditions(merge_point,banned_preconditions):
    '''
    forward check for cycle when a new branch is merged to an existing branch.
    update banned_precondition along its path. If a cycle is detect (a precondition that is banned show up in the existing branch), 
    break the cycle by discarding everything after the actions that requires that precondition (inclusively) and mark preconditions 
    of those actions on the edge as opening conflict
    param merge_point: a precondition where two branchs merge
    param banned_preconditions: banned_preconditions given by the new branch
    '''
    global banned_preconditions_map, knowledge_graph, ends, open_conflict
    ids = []
    for id,node in enumerate(knowledge_graph):
        if id not in ids and merge_point[0]==node[1][0] and merge_point[1]==node[3]:
            ids.append(id)   #actions that require the first common precondition of the two branchs
    pending = []
    for id in ids:
        action = knowledge_graph[id][2].replace('.','')   
        character = knowledge_graph[id][3]
        banned_preconditions_map[(action,character)].extend(banned_preconditions)
        banned_preconditions_map[(action,character)] = list(set(banned_preconditions_map[(action,character)]))
        for node in knowledge_graph:
            if node[3]!=character:
                continue
            if action==node[0][0].replace('.',''):
                precondition = (node[1][0],node[3])
                if precondition in banned_preconditions_map[(action,character)]:
                    for idx,entry in enumerate(knowledge_graph):
                        if entry[2].replace('.','')==action:
                            print("----forward_fix_banned_preconditions: roll back to:",[copy.deepcopy(entry[0]),copy.deepcopy(entry[1])])
                            open_conflict.append((copy.deepcopy(entry[0][0]),copy.deepcopy(entry[0][1]),copy.deepcopy(entry[1][0]),copy.deepcopy(entry[1][1])),copy.deepcopy(entry[3]))
                            knowledge_graph[idx][0] = ("Dummy Sink","Dummy Sink")
                            knowledge_graph[idx][1] = ("Dummy Sink","Dummy Sink")
                            knowledge_graph[idx][2] = "XXXXXXXXXX"
                else:
                    pending.append((precondition,list(set(banned_preconditions_map[(action,node[3])]+[precondition]))))
    for next_precondition,next_banned_precondition in pending:
        #print(next_precondition)
        forward_fix_banned_preconditions(next_precondition,next_banned_precondition)

def calculate_PMI_dc(character, action, hypothesis, pmi_prompt_full, pmi_prompt_domain,question_type='multiQA'):
    '''
    return the PMI support for each hypothesis by formulating it as a multi-choice problem.
    Important Note: PMI trick only works well when the probability of each hypothesis is non-negligible
    param character: string character name
    param action: the action for which certain preconditions need to be satisfied
    param hypothesis: preconditions being ranked
    param pmi_prompt_full: full prompt, contains question, choices, (examples). Read this paper "Surface Form Competition: Why the Highest Probability Answer Isn’t Always Right" for more detail
    param pmi_prompt_domain: usually just "?\nAnswer:"
    '''
    print(action)
    cnt = 0
    PMI_dc = {}
    d = {}    
    pmi_prompt_full_ = str(pmi_prompt_full)
    hypothesis = hypothesis[:8]
    if question_type=='multiQA':
        token_map = {'A':317,'B':347,'C':327,'D':360,'E':412,'F':376,'G':402,'H':367,'I':314}
        option_template = ['{$optionA$}','{$optionB$}','{$optionC$}','{$optionD$}','{$optionE$}','{$optionF$}','{$optionG$}','{$optionH$}','{$optionI$}']
        tokens = list(token_map.keys())
        for i in range(len(hypothesis)):
            PMI_dc[hypothesis[i]]=0
            d[tokens[i]] = hypothesis[i]
            pmi_prompt_full_ = pmi_prompt_full_.replace('{$option'+tokens[i]+'$}', '\n'+tokens[i]+':'+hypothesis[i])
    elif question_type=='TrueFalse':
        token_map = {'Yes':3363,'No':1400,'yes':3763,'no':645}
        option_template = []
        PMI_dc['Yes']=0
        PMI_dc['No']=0
        d['Yes'] = 'Yes'
        d['No'] = 'No'
        d['yes']='Yes'
        d['no']='No'

    #print(d)
    #print(pmi_prompt_full_)
    for template in option_template:
        pmi_prompt_full_ = pmi_prompt_full_.replace(template, '')
    if question_type=='multiQA':
        pmi_prompt_full_ = get_prompt(pmi_prompt_full_, '','',action,character)
    elif question_type=='TrueFalse':
        pmi_prompt_full_ = get_prompt(pmi_prompt_full_, '',' '.join(hypothesis[0].split(' ')[1:]),action,character)    
    #print(pmi_prompt_full_)
    pmi_prompt_domain_ = get_prompt(pmi_prompt_domain)
    _, _, _, logits_full = gpt_call(pmi_prompt_full_, 3, 1, top_p=0.8, top_k=100, return_probability=True)
    _, _, _, logits_domain = gpt_call(pmi_prompt_domain_, 3, 1, top_p=0.8, top_k=100, return_probability=True)
    for k in d.keys():
        PMI_dc[d[k]]+=logits_full[0][0][token_map[k]]/logits_domain[0][0][token_map[k]]
    return PMI_dc


def get_double_negation(sentence, num_generations = 1, verbose = False):
    neg_output = negate_input(sentence, num_generations = num_generations)

    if verbose:
      print("input negation: ", neg_output)

    cfp_gens = counter_fact_precond(neg_output[0][:-1], num_generations = num_generations)

    uniq_cfp_gens, counts = np_unique_sent(cfp_gens, return_counts = True)
    uniq_cfp_gens, counts = list_sort(uniq_cfp_gens, counts)
    uniq_cfp_gens.reverse()
    counts.reverse()

    if verbose:
      print("cfpreconditions: ", uniq_cfp_gens)
      print("counts: ", counts)

    negations = []

    for gen in uniq_cfp_gens:
      all_negations, neg_counts = np_unique_sent(negate_input(gen), return_counts = True)
      all_negations, neg_counts = list_sort(all_negations, neg_counts)

      if verbose:
        print("all candidate negations for ", gen)
        print(all_negations)
        print(neg_counts)

      negations.append(all_negations[np.argmax(neg_counts)])
    
    return negations

def negate_input(sentence, num_generations = 5):
  '''
  return the negation of the input sentence
  param sentence: string. the action or statement
  param generations: int. the number of times to query GPT
  '''

  negation_prompt_step1 = '''Hanna got in the car. The negation of this is Hanna didn't get in the car.
  Steve went to school. The negation of this is Steve didn't go to school.
  Bob gave Jenny a key. The negation of this is Bob didn't give Jenny a key.
  Timmy didn't have a gun. The negation of this is Timmy had a gun.
  {$text1$} The negation of this is'''

  all_gens = []
  prompt = get_prompt(negation_prompt_step1, sentence)

  while len(all_gens) < num_generations:
    gpt_output = gpt_call(prompt, 10, 15, top_p=0.6, top_k=10, return_probability=False)
    output           = gpt_output[0].split("\n")[0]
    # print("output: ", output)
    for atom in split_sentence_array(output):
      if len(atom) > 1 and ("_" not in atom):
        all_gens.append(atom)
  
  return all_gens


def counter_fact_precond(sentence, num_generations = 5):
  '''
  Given a statement, expand on the cause or potential reason behind the statement.
  param sentence: string. the action or statement.
  param generations: int. the number of times to query GPT
  '''

  negation_prompt_step2 = '''Hanna didn't give Bob an apple because Bob was allergic to apples.
  Steve didn't go to the park. This was because the park was closed.
  Jen didn't go into her house. This was because Jen's house was locked.
  Ben didn't go to school. This was because a scary clown was at school.
  Ethan didn't drive to work. This was because Ethan's car was out of gas.
  Jake wasn't hungry. This was because Jake ate a burger.
  {$text1$}. This was because'''

  all_gens = []
  prompt = get_prompt(negation_prompt_step2, sentence)

  max_attempts = 10
  attempts     = 0

  while len(all_gens) < num_generations and attempts < max_attempts:
    gpt_output = gpt_call(prompt, 10, 15, top_p=0.6, top_k=10, return_probability=False)
    output           = gpt_output[0].split("\n")[0]
    # print("output: ", output)
    for atom in split_sentence_array(output):
      if len(atom) > 1 and ("_" not in atom):
        all_gens.append(atom)

    attempts +=1
    
  return all_gens

def list_sort(items, counts, return_reverse = False):
  '''
  Given a list of items and their counts, return the list sorted by counts
  param items: list. an array of unique items
  param counts: list. an array of integers
  '''
  counts, items = (list(t) for t in zip(*sorted(zip(counts, items))))
  
  if return_reverse:
    items.reverse()
    counts.reverse()
    return items, counts
  else:
    return items, counts

def np_unique_sent(array, simi_threshold = .9, return_counts = False):
  '''
  Given a list of statments, find the unique elements and their counts such that no element shares a cosine similarity above a certain threshhold
  param array: list. An array of non-unique sentences.
  param simi_threshold: float. The threshold for the cosine similarity.
  param return_counts: bool. A flag for whether or not to return the counts as well as the unique elements
  '''
  new_array = []
  counts    = []
  for candidate in array:
    cos_score     = 0
    cur_cos_score = 0
    cur_counts    = 1
    for i, element in enumerate(new_array):
      cur_cos_score = cos_simi(candidate, element)
      if cur_cos_score > simi_threshold:
        cos_score   = cur_cos_score
        counts[i]  += 1
        break
      if cos_score < cur_cos_score:
        cos_score   = cur_cos_score

    if cos_score < simi_threshold:
      new_array.append(candidate)
      counts.append(cur_counts)

  if return_counts:
    return new_array, counts
  else:
    return new_array

def get_negation(sentence, character):
    PMI_prompt_full_negation = "Why {$text4$} fails to{$text3$}?'{$optionA$}{$optionB$}{$optionC$}{$optionD$}{$optionE$}{$optionF$}{$optionG$}{$optionH$}{$optionI$}\nAnswer:"
    PMI_prompt_domain_negation = "Why {$text4$} fails to  ?{$optionA$}{$optionB$}{$optionC$}{$optionD$}{$optionE$}{$optionF$}{$optionG$}{$optionH$}{$optionI$}Answer:"
    candidate_negations = call_comet(head=sentence.replace('.','').strip(),character=character)
    candidate_len = len(candidate_negations)
    cnt = 0
    PMI_dc = {}
    d = {}
    hypothesis = candidate_negations[:8]
    token_map = {'A':317,'B':347,'C':327,'D':360,'E':412,'F':376,'G':402,'H':367,'I':314}
    pmi_prompt_full_ = str(PMI_prompt_full_negation)
    pmi_domain_ = str(PMI_prompt_domain_negation)
    for i in range(len(hypothesis)):
        d[chr(ord('A')+i)] = hypothesis[i]
        pmi_prompt_full_ = pmi_prompt_full_.replace('{$option'+chr(ord('A')+i)+'$}', '\n'+chr(ord('A')+i)+':'+hypothesis[i])
    print(d)
    for template in ['{$optionA$}','{$optionB$}','{$optionC$}','{$optionD$}','{$optionE$}','{$optionF$}','{$optionG$}','{$optionH$}','{$optionI$}']:
        pmi_prompt_full_ = pmi_prompt_full_.replace(template, '')
    for i in range(len(hypothesis)):
        pmi_domain_ = pmi_domain_.replace('{$option'+chr(ord('A')+i)+'$}', '\n'+chr(ord('A')+i))
    for template in ['{$optionA$}','{$optionB$}','{$optionC$}','{$optionD$}','{$optionE$}','{$optionF$}','{$optionG$}','{$optionH$}','{$optionI$}']:
        pmi_domain_ = pmi_domain_.replace(template, '')
    pmi_prompt_full_ = get_prompt(pmi_prompt_full_, '','',sentence.replace(character,''),character)
    print(pmi_prompt_full_)
    pmi_prompt_domain_ = get_prompt(pmi_domain_)
    _, _, _, logits_full = gpt_call(pmi_prompt_full_, 3, 1, top_p=0.8, top_k=100, return_probability=True)
    _, _, _, logits_domain = gpt_call(pmi_prompt_domain_, 3, 1, top_p=0.8, top_k=100, return_probability=True)
    for k in d.keys():
        PMI_dc[d[k]]=logits_full[0][0][token_map[k]]/logits_domain[0][0][token_map[k]]
    best = -1
    best_negation = ""
    for k in PMI_dc:
        if PMI_dc[k]>best:
            best = PMI_dc[k]
            best_negation = str(k)
    if len(best_negation)>0:
        return [best_negation]
    else:
        return []


def split_sentence_array(array, end_token = ".", verbose = False):
  '''
  Given a string of sentences, split the sentences into an array and return
  param array: string. The string of sentences to be split
  param end_token: char. the end of sentence token for the string
  '''
  if not array:
    return []
  return_array = array.split(end_token + " ")
  ra = []
  if verbose:
    print("split sentences: ", return_array)

  for element in return_array:
    if element:
      if element[-1] != end_token:
        ra.append(strip_whitespace(element + end_token))
      else:
        ra.append(strip_whitespace(element))

  return ra

def strip_whitespace(string):
  '''
  Given a string, remove all blank spaces from the start and end.
  param stri
  '''
  for i, letter in enumerate(string):
    if letter != " ":
      string = string[i:]
      break
      
  #Reverse string and get other end
  string = string[::-1]
  for i, letter in enumerate(string):
      if letter != " ":
        string = string[i:]
        break

  return string[::-1]

def get_facts_and_check_ownership(sentence, obj, location):
    '''
    return a name of a person who owns the object
    '''
    fact_about_object_prompt = '''Given the sentence, what can be said about the object in the sentence?
Sentence: Tim picked up a bottle of water from the table at home.
What can we tell about the bottle of water?
Answer: There is a bottle of water on the table at Tim's home.

Sentence: Mike read book at the library.
What can we tell about the book?
Answer: There is a book at library.

Sentence: Fred use the microware at kitchen.
What can we tell about the microware?
Answer: There is a microware at Fred's kitchen.

Sentence: Tim bought milk from the grocery store.
What can we tell about milk?
Answer: There is milk for sale at the grocery store.

Sentence: {$text3$}
What can we tell about the {$text4$}?
Answer: There is'''
    ownership_prompt = '''Given the sentence, determine Who owns the object.
Sentence: There is a bottle of water on the table at John's home.
Who owns the bottle of water?
Answer: John.

Sentence: There is milk for sale at the grocery store.
Who owns the milk?
Answer: grocery store.

Sentence:{$text3$}
Who owns the {$text4$}?
Answer:'''
    facts_generation = 'There is'+gpt_call(get_prompt(fact_about_object_prompt,"","",sentence.replace('.','')+' '+location+'.',obj), 8, 1, top_p=0.6, top_k=4)[0].split('\n')[0]
    print("facts about object: ",facts_generation)
    ownership = gpt_call(get_prompt(ownership_prompt,"","",facts_generation.split('.')[0]+'.',obj), 8, 1, top_p=0.6, top_k=4)[0].split('\n')[0].replace('.','').strip()
    return facts_generation,ownership

def get_item_slot_multiple_items(sentence, context='',character=""):
    '''
    fill the slot about what is needed for the character to perform the action. Currently, we restrict it to generate only the most needed item for robustness issue.
    We use few-shot learning to guide GPT to generate 'nothing' if no item is needed. The nothing flag is then removed and will return an empty list of needed items
    Important Note: PMI trick only works well when the probability of each hypothesis is non-negligible. Therefore, we need to filter out outliers first
    param sentence: the action
    param context: string, context
    param character: string
    '''
    check_prompt_needItem = '''{$text1$} needs {$text2$} to {$text3$}  True'''
    check_prompt_becauseNeeds = '''Determine if the following statement is true?\n{$text3$} because {$text1$} needs {$text2$}.  True'''

    #PMI_prompt_full_item_QA = '''What does {$text4$} need to{$text3$}?{$optionA$}{$optionB$}{$optionC$}{$optionD$}{$optionE$}{$optionF$}{$optionG$}{$optionH$}{$optionI$}\nAnswer:'''
    #PMI_prompt_domain_item_QA = '''What does  need to  ?\nA:\nB:\nC\nD\nE\nF\nG\nAnswer:'''
    PMI_prompt_full_item_QA = '''What does {$text4$} need to buy a fishing pole from the fishing store?
A: fishing pole
B: money
Answer: B (money)

What does {$text4$} need to buy a milk from the grocery store?
A: money
B: milk
Answer: A (money)

What does {$text4$} need to{$text3$}?{$optionA$}{$optionB$}{$optionC$}{$optionD$}{$optionE$}{$optionF$}{$optionG$}{$optionH$}{$optionI$}\nAnswer:'''
    PMI_prompt_domain_item_QA = '''What does {$text4$} need to buy a fishing pole from the fishing store?
A: fishing pole
B: money
Answer: B (money)

What does  need to  ?\nA:\nB:\nC\nD\nE\nF\nG\nAnswer:'''
    
    PMI_prompt_full_item_flip = '''money is required to buy milk.
Is this sentence logically correct?
Answer: Yes
milk is required to buy milk.
Is this sentence logically correct?
Answer: No
nothing is required to walk to the post office.
Is this sentence logically correct?
Answer: Yes
{$text2$} is required to {$text3$}.
Is this sentence logically correct?
Answer:'''
    PMI_prompt_domain_item_flip = '''money is required to buy milk.
Is this sentence logically correct?
Answer: Yes
milk is required to buy milk.
Is this sentence logically correct?
Answer: No
nothing is required to walk to the post office.
Is this sentence logically correct?
Answer: Yes
.
Is this sentence logically correct?
Answer:'''
    PMI_prompt_full_true_false_flip = '''Sentence 1: money is needed for buy milk from the grocery store.
Is this statement correct?
Answer: Yes

Sentence 2: ticket is needed for got bus ticket from the ticket office.
Is this statement correct?
Answer: No

Sentence 3: book is needed for read a book.
Is this statement correct?
Answer: Yes

Sentence 4: gun is needed for see bear in the forest.
Is this statement correct?
Answer: No

Sentence 5: computer is needed for the computer broke.
Is this statement correct?
Answer: Yes

Sentence 4: dish cloth is needed for wash dishes.
Is this statement correct?
Answer: Yes

Sentence 5: {$text2$} is needed for {$text3$}.
Is this statement correct?
Answer:'''
    PMI_prompt_full_true_false_flip_domain = '''Sentence 1: money is needed for buy milk from the grocery store.
Is this statement correct?
Answer: Yes

Sentence 2: ticket is needed for got bus ticket from the ticket office.
Is this statement correct?
Answer: No

Sentence 3: book is needed for read a book.
Is this statement correct?
Answer: Yes

Sentence 4: gun is needed for see bear in the forest.
Is this statement correct?
Answer: No

Sentence 5: computer is needed for the computer broke.
Is this statement correct?
Answer: Yes

Sentence 4: dish cloth is needed for wash dishes.
Is this statement correct?
Answer: Yes

Sentence 5:  is needed for .
Is this statement correct?
Answer:'''


    temp = []
    ret = []
    trial_times = [4,1,1]
    trial_top_p = [0.7,0.8,0.8]
    cutoff_frequencies = [0, 1, 1]
    freq_thresholds = [4,1,1]
    for item_cnt in range(3):
        for trial in range(trial_times[item_cnt]):
            text2 = ''
            if len(temp)>0:
                text2 = ''.join([" "+item+',' for item in temp])
            action = ' '.join(sentence.split(' ')[1:]).replace('.','')
            prompt = get_prompt(parallel_precondition_item_prompt,"",text2,action,character)
            #print(prompt)
            
            #filter out outliers first
            output = gpt_call(prompt, 4, 15, top_p=trial_top_p[item_cnt], top_k=50, return_probability=False)
            output = [out for out in output if not any([k in out for k in [' of ',' with ',' as ',' for ', ' to ']])]
            output = post_process(output,prefix="",freq_threshold=4,must_not_contain='no ',match_lemma_threshold=0.2, separator=',', use_majority_vote=True, majority_vote_threshold = 0.7)
            
            if ' nothing' in output:
                output = output[:5]
                if ' nothing' not in output:
                    output.append(' nothing')
            else:
                output = output[:max(3,len(output)//2)]
            output.reverse()
            to_remove = []
            for out in output:
                for out2 in output:
                    if out!=out2 and (out in out2):
                        to_remove.append(out)
            output = [out for out in output if out not in to_remove]
            PMI_dc = calculate_PMI_dc(character,' '+action,output,PMI_prompt_full_item_QA, PMI_prompt_domain_item_QA)
            '''
            PMI_dc = {}
            for candidate in output:
                PMI_dc[candidate] = calculate_PMI_dc(character,action,['has'+candidate],PMI_prompt_full_item_flip,PMI_prompt_domain_item_flip,question_type='TrueFalse')
                print('has'+candidate,PMI_dc[candidate])
                PMI_dc[candidate] = PMI_dc[candidate]['Yes']/PMI_dc[candidate]['No']
            '''
            print(PMI_dc)
            if len(PMI_dc)==0:
                return []
            output = list(PMI_dc.keys())
            probability = [PMI_dc[k] for k in output]
            ret.append(output[probability.index(max(probability))].strip())
            ret = ['has '+r for r in ret if ('nothing' not in r and 'something' not in r)]
            if ret==[]:
                return []
            #PMI_dc_is_item = calculate_PMI_dc('',' '+action,ret,PMI_prompt_full_true_false_flip, PMI_prompt_full_true_false_flip_domain,question_type='TrueFalse')
            #print(ret, PMI_dc_is_item)
            #if PMI_dc_is_item["Yes"]*2.1<=PMI_dc_is_item["No"]:
            #   return []
            #if gpt_call(get_prompt(PMI_prompt_full_true_false_flip,"",ret[0],action,character), 1, 1, top_p=0.5, top_k=1, return_probability=False)[0] == ' Yes':
            #    return ret
            #else:
            #    return []
            
            return ret
        if temp==[]:
            return ret
    return ret


def get_slots_gpt(sentence,context_,last_precondition_type, num_item_state_in_the_plan=0):
    '''
    filling slots (preconditions) given the action. We considered four type of preconditions "item needed", "reason"(why?), "location"(where?), "interaction with other person".
    param sentence: string. the action
    param context_: string. context that explains why the action is take
    '''
    global reasoning_graph
    sentence_ = str(sentence)
    character,characters = find_character_names(sentence)
    characters = [ch.strip() for ch in characters]
    print('*****debug context*****')
    print("Character:",character, characters)
    print(sentence)
    print(context_)
    #context_ = context_hint.replace('{$text1$}',text)
    print('fill slot.. ',sentence)
    slot_flag = try_expand_syntax(get_syntax_prompting(sentence.replace('.','')))
    if len(sentence.split(' '))<4:
        sentence = ' '.join(sentence.split(' ')).replace('.','')+' '+gpt_call(context_+get_prompt(enrich_context_prompt,"",' '.join(sentence.split(' ')).replace('.',''),' '.join(sentence.split(' '))), 13, 1, top_p=0.9, top_k=40)[0].split('\n')[0].strip()
        #sentence = gpt_call(context_+get_prompt(enrich_context_prompt,"","",' '.join(sentence.split(' '))), 13, 1, temperature=0.9)[0].split('\n')[0].strip()
        print("Expand Sentence:",sentence)
    slot = {}
    slot['item needed'] = []
    slot['reason'] = []
    slot['interaction with other person'] = []
    slot['location'] = []
    slot['how']=[]
    #slot_flag['why']=False
    if last_precondition_type=="interaction with other person" or last_precondition_type=='reason':
        slot_flag['negation']=False
    if slot_flag['how']==True:
        how_result = gpt_call(context_+get_prompt(how_to_get_to_goal_prompt,"","",' '.join(sentence.split(' ')).replace('.','')), 8, 15, temperature=0.9)
        #print(how_result)
        how_result = post_process(how_result,remove_first_word=False,mode="gen_actions",to_avoid = sentence,action=sentence, must_have='.', prefix="",separator='.',freq_threshold=15,simi_threshold=0.75)
        if len(how_result)>0:
            how_result = [how_result[0]] if all(k not in how_result[0] for k in [' is ',' was ']) else []
        if len(how_result)>0:    
            slot['how'] =  [(character,how_result[0])]
    #slot_flag['negation']=False
    if slot_flag['negation']==True:
        slot['negation'] = get_negation(sentence_,character)
        slot['item needed'] = []
        slot['reason'] = []
        slot['interaction with other person'] = []
        slot['location'] = []
        return slot_flag, slot, sentence, character
    if last_precondition_type == 'location':
        slot_flag['location']=False
    print(slot_flag)
    if slot_flag['location']==True:
        #print(context_+get_prompt(parallel_precondition_location_prompt,"",character,' '.join(sentence.split(' ')[1:])))
        slot['location'] = []
        for trial in range(4):
            sentence = str(sentence_)
            for person in characters:
                hint_location = comet_call(sentence,5,8,'location',person)
                print("extracted location:",hint_location)
                location=post_process(gpt_call('\n'.join(context_.split('\n')[-3:])+get_prompt(parallel_precondition_location_prompt,"",person,' '.join(sentence.split(' ')[1:]),hint_location), 3, 15, temperature=0.8),prefix="at",freq_threshold=5, use_majority_vote=False)
                #location=post_process(gpt_call(get_prompt(parallel_precondition_location_prompt,"",person,' '.join(sentence.split(' ')[1:])), 3, 15, temperature=0.9),prefix="at",freq_threshold=5, use_majority_vote=False)
                #print('\n'.join(context_.split('\n')[-5:])+get_prompt(parallel_precondition_location_prompt,"",person,' '.join(sentence.split(' ')[1:])),location)
                if location!=[]:
                    location[0] = location[0].replace(' house',' home')
                    if 'where' in location[0] or 'place' in location[0]:
                        break
                    if ' at ' not in sentence and ' from ' not in sentence and person==character:
                        sentence = sentence.split('.')[0].split(' at ')[0]+' '+location[0]
                    if person not in [l[0] for l in slot['location']]:
                        slot['location'].append((person,location[0]))
            if len(slot['location'])==len(characters):
                break
    if slot_flag['item']==True:
        slot['item needed']=get_item_slot_multiple_items(sentence, '', character)
        slot['item needed'] = [(character,item) for item in slot['item needed'] if ' no ' not in item]
        if slot['item needed']!=[] and num_item_state_in_the_plan==0:
            #print(get_prompt(object_state_prompt,"",character,sentence,slot['item needed'][0][1].replace('has','')))
            item_state = gpt_call(get_prompt(object_state_prompt,"",character,sentence,slot['item needed'][0][1].replace('has','')), 12, 1, top_p=0.6, top_k=4)[0].split('\n')[0].replace('.','').strip()
            slot['item state'] = [(character,item_state)] if ('nothing' not in item_state and 'Nothing' not in item_state and slot['item needed'][0][1].replace('has','') in item_state) else []
    if slot_flag['who']==True:
        #print(context_+get_prompt(parallel_precondition_second_character_prompt,"","",' '.join(sentence_.split(' ')).split('.')[0].split(' at ')[0]+' '+slot['location'][0]+'.'))
        for i in range(3):
            if slot_flag['location']==True and slot['location']!=[]:
                who_slot = gpt_call(context_+get_prompt(parallel_precondition_second_character_prompt,"","",' '.join(sentence.split(' ')).split('.')[0]+'.'), 13, 15, top_p=0.5, top_k=40)
                #who_slot = gpt_call(get_prompt(parallel_precondition_second_character_prompt,"","",' '.join(sentence_.split(' ')).split('.')[0].split('at')[0]+slot['location'][0]+'.'), 12, 15, temperature=0.8)
            else:
                who_slot = gpt_call(context_+get_prompt(parallel_precondition_second_character_prompt,"","",' '.join(sentence.split(' '))), 13, 15, top_p=0.5, top_k=40)
                #who_slot = gpt_call(get_prompt(parallel_precondition_second_character_prompt,"","",' '.join(sentence_.split(' '))), 12, 15, temperature=0.8)
            #print("debug who :",who_slot)
            slot['interaction with other person']=post_process(who_slot,prefix="",separator=';',must_not_contain = ')',
                                                    remove_first_word=False,mode="gen_actions",to_avoid = ' '.join(sentence.split(' ')),lookup={},simi_threshold=0.95,freq_threshold=15)
            if slot['interaction with other person']!=[]:
                print("debug interaction with other person :",slot['interaction with other person'])
                #slot['interaction with other person'][0]=slot['interaction with other person'][0]+' '+slot['location'][0]
                #slot['interaction with other person'] = [interaction.strip() for interaction in slot['interaction with other person'][0].split('.')]
                slot['interaction with other person'] = [(character,interaction) for interaction in slot['interaction with other person'] if len(interaction)>0]
                break
    if slot_flag['why']==True:
        reason = gpt_call(context_+get_prompt(parallel_reason_prompt,"","",' '.join(sentence.split(' ')).replace('.','')), 8, 15, temperature=0.9)
        slot['reason']=post_process(reason,remove_first_word=False,mode="gen_actions",to_avoid = sentence,action=sentence, must_not_contain = 'of ',
                                    must_have='.', prefix="",separator='.',freq_threshold=15,simi_threshold=0.6)
        if len(slot['reason'])>0:
            slot['reason'] = [slot['reason'][0]] if all([k not in slot['reason'][0] for k in [' is ',' was ']]) else []
        if len(slot['reason'])>0:
            reasoning_graph.append([sentence_.strip().replace('.',''),slot['reason'][0]])
            slot['reason'] = [(character,slot['reason'][0])]
        #yield reason[0].strip(),"To "+end+", the person need to "+reason[0]+", which is solved by "+reason[0]+".\n",already_satisfied_preconditions
    return slot_flag, slot, sentence, character

def gpt_planner_single_step(end,lookup={},initial_state=[]):
    '''
    See the documentation at the beginning of this cell
    data struction:
        knowledge_graph: list[list[tuple(action, elaboration of action), tuple(precondtion, type of precondition), solution, character name]] 
        banned_preconditions_map: map{tuple(string,string)->list[tuple(string)]} [tuple(action,character):list[tuple(banned precondtions,character)]]
        knowledge_graph: list [tuple(action, elaboration of action), tuple(precondtion, type of precondition), solution, character name]
        open_conflict: list[list[action, elaboration of action, precondtion, type of precondition, character name]] 
        reasoning_graph: list[list[event, reason]]
        reusable_solutions: map{string->list[string]} [precondition: list of buffered candidate solutions]
    param end: the goal sentence
    param lookup: not used yet
    initial_state: list of string. Each item in initial_state is a clause that is satisfied for every characters by default. Example like "has money", "has car". It is used to end the recursion.
    '''
    global knowledge_graph, banned_preconditions_map, ends, open_conflict,reasoning_graph,reusable_solutions
    open_conflict=[]
    end_ = str(end)
    character = find_character_names(end)[0]

    #backtrack to produce context_ that help explain why current action is taken, also maintain banned_preconditions_map
    backtrack_result, num_item_state_in_the_plan = backtrack(end_.strip(),character)
    context_  = context_hint.replace('{$text1$}','\n'.join(['\n'.join([event[0]+' because '+event[1] for event in reasoning_graph if ('want' in event or 'need' in event)]),backtrack_result]))
    last_precondition_type = None
    last_action_item_needed = None
    last_action = None
    last_precondition = ''
    for entry in knowledge_graph:
        if end==entry[2]:
            last_precondition_type = entry[1][1]
            last_precondition = entry[1][0]
            last_action = entry[0][0]
    for entry in knowledge_graph: 
        if last_action==entry[0][0] and entry[1][1]=='item needed':
            last_action_item_needed = entry[1][0]

    #Fill slots (candidate preconditions)
    slot_flag, slots_filtered, end, character = get_slots_gpt(end,context_,last_precondition_type, num_item_state_in_the_plan=num_item_state_in_the_plan)

    if last_precondition_type=='item state' and last_action_item_needed:
        slots_filtered['item needed'].append((character,last_action_item_needed))
    #initialize already_satisfied_preconditions
    already_satisfied_preconditions = [(str(entry),1,character) for entry in initial_state]
    for entry in knowledge_graph:
        # we don't want the same character to perform the same action twice in a plan as this is a greedy planner
        already_satisfied_preconditions.extend([(entry[0][0],0,entry[3]),(entry[0][1],0,entry[3]),(entry[1][0],1,entry[3]),(entry[2],0,entry[3])])  
    already_satisfied_preconditions = list(set(already_satisfied_preconditions))
    print("iteration goal: %s"%end)

    #what preconditions are not allowed to appear for currect node. banned_preconditions_map is maintained by backtrack
    banned_preconditions = banned_preconditions_map[(end_.replace('.',''),character)]
    print('banned_preconditions: ',banned_preconditions)
    
    print("Elaborating:",end)
    print(slots_filtered)
    candidate_preconditions = []
    for key in slots_filtered:
        candidate_preconditions.extend([(precondition,key) for precondition in slots_filtered[key]])
    if len(candidate_preconditions)==0:
        "No action is selected"," finished"
    print("-----candidate_preconditions:")
    print(candidate_preconditions)

    #find open preconditions. Part I
    #detect conflict
    no_conflict_flag = True
    fact_about_object = ''
    ownership_of_object = ''
    if last_precondition_type=='item needed' and len(slots_filtered['location'])>0:
        fact_about_object,ownership_of_object = get_facts_and_check_ownership(end_,last_precondition.replace('has ',''),slots_filtered['location'][0][1])
    for (precondition,precondition_type) in copy.deepcopy(candidate_preconditions):
        no_conflict_flag = True
        if precondition_type=='negation':
            no_conflict_flag = False
        for banned_precondition in banned_preconditions:
            #print("debug banned precondition:",banned_precondition,precondition)
            if banned_precondition[-1]!=precondition[0]:
                continue
            similarity = cos_simi(precondition[1],banned_precondition[0])
            if precondition_type=='item needed' or precondition_type=='location':
                if banned_precondition[0].split(' ')[0]!=precondition[1].split(' ')[0]:
                    continue
                if similarity>0.75:
                    if last_precondition_type=='item needed' and cos_simi(precondition[1],last_precondition)>0.75:
                        #We recover some actions where the item belongs to the character because we want actions like "John get a book from the shelf at John's home" in our plan
                        #this helps to boost diversity.
                        if ownership_of_object==character:
                            print("**recovered a deletion based on facts")
                            print(end_,ownership_of_object,fact_about_object,precondition,last_precondition)
                            if (precondition,precondition_type) in candidate_preconditions:
                                candidate_preconditions.remove((precondition,precondition_type))
                            for entry_precondition in candidate_preconditions:
                                if entry_precondition[1]=='item state':
                                    candidate_preconditions.remove(entry_precondition)
                            break
                    print("delete banned precondition 1: ",precondition,banned_precondition[0],similarity)
                    no_conflict_flag = False
                    break
            else:
                if precondition_type == 'interaction with other person' and similarity>0.8:
                    if (precondition,precondition_type) in candidate_preconditions:
                        candidate_preconditions.remove((precondition,precondition_type))
                elif similarity>0.8:
                    print("delete banned precondition 2: ",precondition,banned_precondition[0],similarity)
                    no_conflict_flag = False
                    break
        if no_conflict_flag==False and 'want' not in precondition[1] and 'need' not in precondition[1]:
            #if action is banned, we delect this action (entire node), gather its precondtions, find another way to satisfy the precondtion
            candidate_preconditions = []
            #delete previous action, collect entry from knowledge graph that starts with this action and collect entries whose solution is this action, add unsolved precondition caused by the deletion to the pending Q.
            index_to_remove = [idx for idx,entry in enumerate(knowledge_graph) if entry[0][0]==end_]
            temp = copy.deepcopy(knowledge_graph)
            knowledge_graph = [copy.deepcopy(temp[i]) for i in range(len(temp)) if i not in index_to_remove]
            reasoning_graph = [reason for reason in reasoning_graph if (end_.strip().replace('.','')!=reason[0] and end_.strip().replace('.','')!=reason[1])]
            index_to_pend_and_remove = [idx for idx,entry in enumerate(knowledge_graph) if entry[2]==end_]
            new_precondition_unsolved = [(knowledge_graph[i][0][0],knowledge_graph[i][0][1],knowledge_graph[i][1][0],knowledge_graph[i][1][1],
                                          knowledge_graph[i][3]) for i in index_to_pend_and_remove]
            open_conflict.extend(new_precondition_unsolved) 
            for i in range(len(new_precondition_unsolved)):
                print("remove similar solution from reusable_solutions",new_precondition_unsolved[i][2],end_[5:].strip().replace('.',''))
                to_remove_action = end_.strip().replace('.','')
                if 'has ' in new_precondition_unsolved[i][2] or 'at ' in new_precondition_unsolved[i][2]:
                    to_remove_action = ' '.join(to_remove_action.split(' ')[1:])
                reusable_solutions[(new_precondition_unsolved[i][2].strip(),new_precondition_unsolved[i][4])] = [solution for solution in reusable_solutions[(new_precondition_unsolved[i][2].strip(),new_precondition_unsolved[i][4])] if cos_simi(solution, to_remove_action)<0.75]
            temp = copy.deepcopy(knowledge_graph)
            if precondition_type!='negation':
                knowledge_graph = [copy.deepcopy(temp[i]) for i in range(len(temp)) if i not in index_to_pend_and_remove]
            else:
                print("add negation",precondition)
                knowledge_graph.append([(end_,end),(precondition,'negation'),'which is satisfied automatically.',character])

            print("add unsolved precondition: ",new_precondition_unsolved)
            break
         
    #print("-----already_satisfied_preconditions:")
    #print(already_satisfied_preconditions)
    
    for precondition,precondition_type in candidate_preconditions:
        no_match_flag = True
        to_merge_node = None
        for already_satisfied_precondition in already_satisfied_preconditions:
            if already_satisfied_precondition[-1]!=precondition[0]:
                continue
            similarity = cos_simi(precondition[1],already_satisfied_precondition[0])
            #merge duplicate precondition
            if 'has '==precondition[1][:4] or 'at '==precondition[1][:3]:
                if similarity>0.75:
                    print("merge duplicate precondition 1: ",precondition[1],already_satisfied_precondition[0],similarity)
                    no_match_flag = False
                    to_merge_node = copy.deepcopy(already_satisfied_precondition)
                    break
            else:
                if similarity>0.8:
                    print("merge duplicate precondition 2: ",precondition[1],already_satisfied_precondition[0],similarity)
                    no_match_flag = False
                    to_merge_node = copy.deepcopy(already_satisfied_precondition)
                    break
        if no_match_flag and 'want' not in precondition[1] and ('need' not in precondition[1] or precondition_type=='item state'):
            open_conflict.append((end_,end,precondition[1],precondition_type,precondition[0]))
        else:
            #fix conflict caused by merging
            #forward_fix_banned_preconditions(precondition[1],list(set(banned_preconditions+[(precondition[1],precondition[0])])))
            if to_merge_node:
                knowledge_graph.append([(end_,end),(to_merge_node[0],precondition_type),'which is satisfied automatically.',precondition[0]])
            else:
                knowledge_graph.append([(end_,end),(precondition[1],precondition_type),'which is satisfied automatically.',precondition[0]])
            yield "","","",True
    
    
    if fact_about_object!='' and len(candidate_preconditions)>0:
        knowledge_graph.append([(end_,end),(fact_about_object,'fact_about_object'),'which is satisfied automatically.',character])
    '''if len(candidate_preconditions)>0 and (last_precondition_type=='item needed' or last_precondition_type=='location'):
        double_negation = get_double_negation(end_)[0]
        knowledge_graph.append([(end_,end),(double_negation,'fact_about_object'),'which is satisfied automatically.',character])'''
    open_conflict = list(set(open_conflict))
    print("open conflict:",open_conflict)

    #knowledge_graph = [entry for entry in knowledge_graph if entry[1][1]!='negation' or cos_simi(entry[0][0],end_)<0.7]

    #find next actions (solutions) to take given open preconditions
    #actions for "item needed" are buffered and will be reused if the action is delected for causing a conflict
    while len(open_conflict)!=0:
        original_sentence, expanded_sentence,selected_precondition,selected_precondition_type, character_  = open_conflict[0]
        open_conflict.pop(0)
        print("selected_precondition for (%s):"%original_sentence)
        print(selected_precondition+' to '+' '.join(original_sentence.split(' ')[1:]))
        candidate_actions_filtered = []
        #print(reusable_solutions,(selected_precondition.strip(), character_))
        print("buffered solution for "+character_+" "+selected_precondition+":",reusable_solutions[(selected_precondition.strip(), character_)])
        if len(reusable_solutions[(selected_precondition.strip(), character_)])>0:
            if selected_precondition_type=='interaction with other person':
                continue
            candidate_actions_filtered = copy.deepcopy(reusable_solutions[(selected_precondition.strip(), character_)])
        else:
            satisfy_flag=False
            print("-----candidate_actions:")
            while satisfy_flag==False:
                if selected_precondition_type=='item needed' or selected_precondition_type=='location' or selected_precondition_type=='item state':
                    if selected_precondition_type=='item needed':
                        gpt_generated_actions = gpt_call(get_prompt(action_prompt_obtain_item,' '.join(expanded_sentence.split(' ')[0:]).strip(),character_,preprocess(selected_precondition+'?')),8, 20, temperature=0.9)
                    if selected_precondition_type=='location':
                        if any([key_word in selected_precondition for key_word in ['stand','station','stop']]) and any([key_word in selected_precondition for key_word in ['bus','taxi','subway']]):
                            gpt_generated_actions = [character_+' walked to get to '+selected_precondition.replace('at ','')+'.',character_+' drove to get to '+selected_precondition.replace('at ','')+'.']
                        else:
                            gpt_generated_actions = gpt_call(get_prompt(action_prompt_go_to_somewhere,' '.join(original_sentence.split(' ')[0:]).strip(),character_,preprocess(selected_precondition+'?')),9, 5, top_k=7, top_p=0.6)

                    if selected_precondition_type=='item state':
                        gpt_generated_actions = gpt_call(get_prompt(object_state_to_action_prompt,"",character_,original_sentence,selected_precondition),12, 5, temperature=0.75)
                    candidate_actions = post_process(gpt_generated_actions,'.',
                                        mode="gen_actions",must_have='.', to_avoid = ' '.join(original_sentence.split(' ')[1:]),lookup=lookup,action='', simi_threshold=0.8,freq_threshold=0,exclude_go_to=not ' at ' in selected_precondition)
                
                    if 'at ' in selected_precondition:
                        candidate_actions = [action.split(' to ')[0]+' to get to '+' '.join(selected_precondition.split(' ')[1:]) for action in candidate_actions]
                else:
                    candidate_actions = [selected_precondition]
        
                if len(candidate_actions)==0:
                    continue
            
                candidate_actions_filtered = []
                print("debug candidate_actions:",candidate_actions,selected_precondition)
                for candidate_action in candidate_actions:
                    if character_ != candidate_action.strip().split(' ')[0]:
                        candidate_action = character_+' '+candidate_action.strip()
                    if selected_precondition[0:3]=='has': 
                        candidate_actions_filtered.append(fix_missing_object(candidate_action,selected_precondition[4:]))
                    elif selected_precondition[0:2]=='at':
                        candidate_actions_filtered.append(fix_missing_object(candidate_action,selected_precondition[3:]))
                    else:
                        candidate_actions_filtered.append(candidate_action)

                candidate_actions_filtered = choose_best_action_by_perplexity(selected_precondition,candidate_actions_filtered,expanded_sentence,character_)[1]
                
                if len(candidate_actions_filtered)==0:
                    continue
                else:
                    satisfy_flag = True
                    break

            reusable_solutions[(selected_precondition.strip(), character_)] = copy.deepcopy(candidate_actions_filtered)

        print("candidate_actions_filtered:",candidate_actions_filtered)
        #selected_action = choose_best_action_by_perplexity(selected_precondition,candidate_actions_filtered).strip()
        selected_action = candidate_actions_filtered[0] #choose_best_action_by_perplexity(selected_precondition,actions,parent_action,character)[0]
        selected_action = selected_action.strip()
        print("selected_action for (%s):"%selected_precondition)
        if selected_precondition_type=='interaction with other person':
            character_ = find_character_names(selected_action)[0]
        elif character_!=selected_action.split(' ')[0]:
            selected_action = character_+' '+selected_action
        selected_action = selected_action.replace('.','')
        print(selected_action)
        knowledge_graph.append([(original_sentence,expanded_sentence),(selected_precondition,selected_precondition_type),selected_action+'.',character_])
        yield selected_action, character_,selected_precondition_type,no_conflict_flag

def iterative_gpt_planner(end,iterations=10,initial_state=['has car']):
    '''
    See the documentation at the beginning of this cell
    param end: the goal sentence
    param iterations: how many iterations until we cut the recursion
    initial_state: list of string. Each item in initial_state is a clause that is satisfied for every characters by default. Example like "has money", "has car". It is used to end the recursion.
    '''
    global text, banned_preconditions_map, knowledge_graph, reasoning_graph,ends,reusable_solutions
    reusable_solutions = defaultdict(list)
    text = ""
    end_ = str(end)
    ends = [end]
    banned_preconditions_map = defaultdict(list)
    banned_preconditions_map[end]=[]
    knowledge_graph = []
    reasoning_graph = []
    terminate = False
    for iter in range(iterations):
        print("knowledge graph:")
        print(knowledge_graph)
        new_ends = []
        time_out = 0
        while len(ends)!=0:
            print('****edges: ',ends)
            end = ends.pop(0)
            planner = gpt_planner_single_step(end,initial_state=initial_state)
            out = list(planner)
            if len(out)==0:
                continue
            for o in out:
                if o[-1]==False:
                    if time_out<=5:
                        if o[2]=='interaction with other person':
                            ends.insert(0,(o[0]+'.').strip())
                        else:
                            if  o[1]+' ' not in o[0]:
                                ends.insert(0,(o[1]+" "+o[0]+'.').strip())
                            else:
                                ends.insert(0,(o[0]+'.').strip())
                        time_out+=1
                        print("----------------try to fix cycle: %i--------------"%time_out)
                    else:
                        knowledge_graph = [node for node in knowledge_graph if node[0][0]!=end]
                        time_out=0
                elif len(o[0])>0:
                    new_ends.append(o)
        ends = []
        print("--------Iteration %i--------"%iter)
        for action,character,action_type,no_conflict_flag in new_ends:
            if len(action)!=0:
                if action_type=='interaction with other person':
                    ends.append((action+'.').strip())
                else:
                    if  character+' ' not in action:
                        ends.append((character+" "+action+'.').strip())
                    else:
                        ends.append((action+'.').strip())
        ends = list(set(ends))
        print(ends)
        if len(ends)==0:
            break
        if terminate:
            break
    text = forward_traverse(end_)
    print("banned precondition map:")
    print(banned_preconditions_map)
    print("full text:")
    print(text)
    print("\n\nfinal text:")
    [print(s) for s in text.split('\n') if 'automatic' not in s]
    return text.split('\n'),copy.deepcopy(knowledge_graph)
#iterative_gpt_planner("John gave Sally a book.",iterations=5)
```
